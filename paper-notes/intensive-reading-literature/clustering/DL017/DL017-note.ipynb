### 深度表示和图像聚类的联合无监督学习  

<br>
  
  
##### 要解决的问题      
  
  用循环神经网络做无监督的图像聚类；  
    
  在统一的框架中利用无监督的图片同时做图像聚类和表示学习（降维吗？）  

<br>  
  
##### 思想  
    
  聚类算法的连续操作变成了这个框架中的循环神经网络。。  
    
  图像聚类和表示学习同时更新；  
    
  图像聚类前向进行；表示学习学习后向进行  
     
  思想是：后向学习得到好的表示学习有利于图像聚类，前向聚类结果为表示学习提供监督信号。  
    
  把这两个过程统一到权重的三元组损失函数；优化方法使用端到端优化   
    
      
 <br>  
大体流程：  

给定ns个无标记原始图片：$I=\{I_1,I_2,\cdots,I_{ns}\}$:  
  
全局损失函数(y是聚类ID,$\theta$是表示学习的参数)：$\underset{y,\theta}{arg\;min } L(y,\theta|I)$  
  
固定$(y,\theta)$中一个，损失函数的优化就可以迭代优化  

$\underset{y}{arg\;min } L(y|\theta，I)    \qquad (2a)$   
  
$\underset{\theta}{arg\;min } L(\theta|I，y)  \qquad (2b)$ 

     
     
<br>  
  
##### 使用的方法       
       
 ns个无标记原始图片：$I=\{I_1,I_2,\cdots,I_{ns}\}$  ;初始聚类ID: $Y=\{y_1,y_2,\cdots,y_{ns}\} \quad$ CNN网络的参数$\theta$ ;基于这个CNN 可以把图像学的特征x : $\quad I=\{I_1,I_2,\cdots,I_{ns}\} \;\; \Rightarrow X=\{x_1,x_2,\cdots,x_{ns}\}$  
   
给出图像的预测标签，可以把这些图片聚为nc个类$C=\{c_1,c_2,\cdots,c_{nc}\} \qquad\quad C_i=\{x_k|y_k=i  \} \quad k \in (1,2,……，ns)$  
    
<br>    
    
$N_i^{k_s}$表示 $x_i$的$K_s$个近邻；$N_{c_i}^{k_c}$表示聚类 $c_i$的$K_c$个近邻  
  
<br>   
对$N_{c_i}^{k_c}$中，按与聚类 $c_i$的亲和度进行降序排列，这样$\underset{x\in C^t}{arg\;max}{A(C_i,C)}$是这个集合中第一个$N_{c_i}^{k_c}$
         
    
<br>   
  
测量两个聚类的亲和度的函数是：$A(.)$

<br>  
聚类方法采用的是凝聚聚类：  
  
  每次合并最相似的两个聚类；这个最相似用亲和度来衡量：$\{c_a,c_b\}=\underset{c_i \neq c_j}{arg\;max} \;A(c_i,c_j)$  
  
<br>  
  
    
 循环框架（感觉上应该主体应该是循环神经网络）：  
 
   CNN网络+单次聚类两个最相似簇==循环单元

每一次聚类（时间步t）:  

1、图片集I进入CNN,学的t步表征xt;作为t步的输入；其中CNN的监督信号来自上一个时间步得到的聚类标签  

2、xt表征利用亲和度矩阵；来合并最相似的两个簇；得到这个时间步的隐藏单元（聚类标签）；用于下一个时间步的CNN(后向)的监督信号
