### 深度卷积GAN（DCGAN）  

在本文中，作者做出了以下贡献：  
  
1、我们提出并评估了卷积 GAN 的体系结构拓扑上的一组约束，这些约束使它们在大多数情况下都能稳定地进行训练。我们将这类架构命名为Deep Convolutional GANs (DCGAN)  
  
2、我们使用训练过的识别器进行图像分类任务，显示出与其他无监督算法的竞争性能  
  
3、我们将GANs学习到的过滤器可视化，并通过经验表明，特定的过滤器已经学会绘制特定的对象  
    
      
4、我们展示了生成器具有有趣的向量算术属性，可以方便地操作生成的示例的许多语义质量

### 方法和模型结构    
  
我们的方法的核心是采用和修改CNN架构最近演示的三个更改 
  
  1、第一个是all convolutional net (Springenberg et al.， 2014)，它将确定性的空间池函数(如maxpooling)替换为strided convolutions，允许网络学习自己的空间下采样。我们在生成器中使用这种方法，允许它学习自己的空间上采样和鉴别器  
    
      
 2、GAN的第一层以均匀噪声分布Z作为输入，可以称为全连通，因为它只是一个矩阵乘法，但结果被重新构造成一个四维张量，用作卷积堆栈的起点。对于鉴别器，将最后一个卷积层扁平化，然后输入单个sigmoid输出。  
   
3、批量标准化(Ioffe & Szegedy, 2015)，通过将每个单元的输入归一化，使其均值和单位方差为零，从而稳定学习。这有助于处理由于糟糕的初始化而出现的训练问题，并有助于在更深的模型中实现梯度流。这被证明是让深层生成器开始学习的关键，防止生成器将所有样本崩溃到一个单点，这是在GANs中观察到的常见故障模式。然而，将batchnorm直接应用于所有层，会导致样本振荡和模型不稳定。这是通过不应用batchnorm到生成器输出层和鉴别器输入层来避免的。  除输出层使用Tanh函数外，在生成器中使用ReLU激活(Nair & Hinton, 2010)。我们观察到，使用有界激活使模型能够更快地学习，以饱和和覆盖训练分布的颜色空间。在鉴别器中，我们发现LekRelu

  稳定的深卷积GANs的架构指南： 
    
1、用strided convolutions (discriminator)和fractionstrided convolutions (generator)替换掉所有池化层    
  
  
  
2、在生成器和鉴别器中使用batchnorm（批标准化）    
  
3、为更深层的架构移除完全连接的隐藏层  
  
4、除使用Tanh的输出外，在生成器中对所有层使用ReLU激活   
  
5、在鉴别器中对所有层使用LeakyReLU激活
