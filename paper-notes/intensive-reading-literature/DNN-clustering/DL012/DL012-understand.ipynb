### 迈向K均值友好空间：同时进行深度学习和聚类
在做降维时；需要考虑的事情：1、假如在未降维时又很好的聚类结构，但做了降维后失去聚类结构；2、降维的数表示形式又很多种，如何保证我们做的降维，后数据又很好的聚类结构。  
  
  这篇文章讲到最近的研究表明，同时做这两项工作，可以很大程度上解决上面的问题。我的理解就是在做降维的时候，同时做聚类任务，就是选择聚类结构好的那一种降维形式。  
    
      
 ##### 局限性  
 DL019就是同时做降维和聚类，但是他是通过线性的降维，从复杂的转化到重构。文章提出了一种DR和K-means聚类联合方法，其中DR是通过学习深度神经网络（DNN）来完成的。 这样做的动机是保留联合优化两个任务的优势，同时利用深层神经网络近似任何非线性函数的能力
