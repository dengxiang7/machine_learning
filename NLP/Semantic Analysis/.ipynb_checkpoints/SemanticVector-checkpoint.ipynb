# 语义向量
&emsp;&emsp; 主题向量、词向量、语义向量三种不同的叫法实质上都是用来表示词或者文档背后的语义。（机器只能计算意义，而不能理解意义。） 

#### 主题向量
&emsp;&emsp; TF-IDF向量的缺陷：TF-IDF向量是对文档中准确拼写的词形进行计数。因此，如果表达相同含义的文本使用了词的不同拼写方式将得到完全不同的TF-IDF向量。因此，基于词统计的应用将变得不可靠。（基于语义的应用则不同。）即使使用了词形归并和词干还原来创建拼写相似、含义也通常相似的小型词集合也不能解决问题，词形归并将相似拼写的词放在一起，但这些词可能意义不同。基于此，$\color{#F00}{在我们得到的TF-IDF向量空间中，如果两段内容相同， 但是使用不}$$\color{#F00}{同的词，那么他们在此空间中不会接近，而有时，两个词形归并后TF-IDF向量虽然接近，但在意义上可能完全不相似。}$  
  
&emsp;&emsp; 主题向量种类：词含义的向量称为词-主题向量，文档含义的向量称为文档-主题向量，这两种向量都是主题向量，只是表示的对象不同。（词库中的每个词都有一个词-主题向量，对于一篇新文档，可以用语料库中的词-主题向量计算出文档-主题向量）   
  
&emsp;&emsp; 主题向量的挑战：对于词和句子的语义进行数值化非常棘手。1、一词多义 2、同音异义 3、轭式搭配  4、同形异义  5、同音异形  
  
&emsp;&emsp; 主题向量的思想： 找到属于同一个主题的那些词维度，然后对这词维度的TF-IDF值求和，以创建$\color{#F00}{一个新的数值 }$来表示文档中$\color{#F00}{该主题的权重}$。甚至可以对此维度进行加权以衡量他们对主题的重要度。以及我们所希望的每个词对这个组合的贡献度。我们也可以用负权重表示词，从而降低文本与该主题相关的可能性。  
  
&emsp;&emsp; 文档-主题向量的做法： 人类如何从数学上确定哪些主题和词相互关联，以及这些关联的权重。例如；确定了3个要建模的主题，就必须确定这些主题中的每一个词的权重（我们让每一个主题都按比例混合词；即；权重* 词项的tf-idf）。主题建模是一个3 * 6的矩阵（权重），代表3个主题与6个词项相关（每个主题有这些6个词项的加权表示），用这个矩阵乘以一个假想的$6\times 1$的TF-IDF向量（tf-idf），就可以得到该文档的$3\times 1$主题向量。（我们如何对这个矩阵赋值，每个人头脑中的语料库不同，对这个矩阵会赋不同的值，如何达到共识。）  
  
&emsp;&emsp; 词-主题向量的做法：对文档-主题向量进行转置就可以得到词-主题向量。（每个词的语义）  
  
对主题向量矩阵的赋值使用人工赋值时主观的人力耗费型的语义分析方法以来人们的直觉和常识来将文档解析为主题（常识很难编码到算法中）。
