### AP 聚类  
  
  通过在数据点之间传递消息进行聚类   
    
#### 摘要
  
  通过识别具有代表性的样本子集对数据进行聚类，对于处理感知信号和检测数据中的模式具有重要意义。这种“范例”可以通过随机选择一个初始数据点子集，然后迭代地精炼它来找到，但是只有当初始选择接近一个好的解决方案时，这种方法才有效。我们设计了一种称为“关联传播”的方法，它将数据点对之间的相似性作为输入度量。实值消息在数据点之间交换，直到出现一组高质量的范例和相应的集群。我们使用亲和传播对人脸图像进行聚类，检测微阵列数据中的基因，识别手稿中有代表性的句子，以及识别通过航空旅行有效访问的城市。关联传播发现集群的错误要比其他方法低得多，并且它在不到百分之一的时间内就做到了这一点  
    
  
#### 引言  
  
  基于相似性度量的数据聚类是科学数据分析和工程系统中的关键步骤。一种常见的方法是使用数据来学习一组中心，使数据点与其最近的中心之间的误差平方和很小。当从实际数据点选择中心时，它们被称为“范例”。“流行的k-centers聚类技术(1)从一组随机选择的样本开始，迭代地改进这组样本，以减少平方和误差。k-centers集群对范例的初始选择非常敏感，因此它通常会多次使用不同的初始化来重新运行，以尝试找到一个好的解决方案。但是，只有当集群的数量很少并且很有可能至少有一个随机初始化接近于一个好的解决方案时，这种方法才能很好地工作。我们采用了一种完全不同的方法，并引入了一种同时考虑所有数据点作为潜在范例的方法。通过将每个数据点视为网络中的一个节点，我们设计了一种沿网络边缘递归传输实值消息的方法，直到出现一组好的范例和相应的集群。如后面所述，消息是在搜索适当选择的能量函数的最小值的简单公式的基础上更新的。在任何时间点，每个消息的大小反映了一个数据点在选择另一个数据点作为其范例时的当前关联性，因此我们将我们的方法称为“关联性传播”。图1A说明了在消息传递过程中集群是如何逐渐出现的。

<br>  
关联传播并不要求预先指定集群的数量，而是将每个数据点k的实数s(k,k)作为输入，这样具有较大s(k,k)值的数据点更有可能被选择为范例。这些值称为“首选项”。识别范例的数量(集群的数量)受到输入首选项的值的影响，但也来自于消息传递过程。如果是先验的，所有数据点都同样适合作为范例，那么应该将首选项设置为一个公共值——这个值可以进行更改，以生成不同数量的集群。共享值可以是输入相似性的中位数(导致集群数量适中)，也可以是输入相似性的最小值(导致集群数量较少)。
