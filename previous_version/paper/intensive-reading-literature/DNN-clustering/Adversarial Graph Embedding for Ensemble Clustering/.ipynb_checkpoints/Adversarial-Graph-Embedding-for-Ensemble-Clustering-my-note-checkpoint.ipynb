# 用于集成聚类的对抗图嵌入   
  
本模型（AGCN）为：是一种聚类集成的方法，模型的输入一致性图信息（关联矩阵）。（拓扑图），使用GCN提取这个拓扑的空间特征z，利用这个空间特征z,进行对抗性训练。
  
网络结构：   
  
  ![](img/1.PNG)  
    
<br>  
  
### **部分模型：GCN**  
    
   用来提取输入数据是拓扑图，的空间特征。  
     
   GCN也是一个神经网络层，它的层与层之间的传播方式是：  
  
 $$Z^{(l+1)}=\sigma({\hat{D}}^{1/2}\hat{A}\hat{D}^{1/2})Z^{(l)}W^{(l)}$$  
   
其中Z是每层获得的特征。（输入层时是：X）即卷积结果   
   
$\hat{A}=A+I$，I是单位矩阵   
  
$\hat{D}$是$\hat{A}$的度矩阵（degree matrix）  
  
σ是非线性激活函数  
  
W是每层的滤波参数矩阵   

  
<br>  


#### **内积层：** 用来重构关联矩阵（保证提取的特征有效）  
  
   $$P(S|Z)=\prod_{i=1}^N \prod_{j=1}^N S(Z_iZ_j^T)$$   
     
<br>  

<br>   
  
#### **GCN的重建：**  
  
  GCN提取空间特征Z：  
  
 $$q_{\phi}(Z|X,S)=\prod_{i=1}^N q_{\phi}(Z_i|X,S)$$  
   
  基于Z重建关联矩阵S:  
  
 $$P(S|Z)=\prod_{i=1}^N \prod_{j=1}^N S(Z_iZ_j^T)$$   
   
则GCN的重建损失为（交叉熵损失）：  
  
 $$\underset{q_{\phi}}{min} Loss=E_{q_{\phi}(Z|X,S)}[-log P(S|Z)]$$ 
    

<br>

有了特征之后，我们利用这个特征进行对抗性训练。  
  
<br>   

<br> 


### **识别器（Discriminate）：** 用来识别是先验分布产生的特征，还是GCN产生的特征  
    
$$D=f(z;\psi)$$  
   
<br>  

#### **对抗训练中：先验分布（假定的真实数据所服从的分布）**  
  
 $$P(Z|H)=\sum_{c=1}^K P(Z|c)P(c|H)$$  
  
这里$P(Z|c)=N(u_c,\sigma_cI) \qquad P(c|H)=cat(c|\pi)$  
    
 $$H_{ik}=\frac{(1+||Z_i-\theta_k||^2/\alpha)^{-\frac{\alpha+1}{2}}}{\sum_{i=1}^k(1+||Z_i-\theta_k||^2/\alpha)^{-\frac{\alpha+1}{2}}}$$   
   
 先验分布：根据Z计算属于哪个类别C，在从这个类被中采样特征Z.(要生成的Z接近真实类别对应的特征Z)    
   
<br>  
  
<br>  

所以对抗训练的目标函数是：  
  
  最大识别先验分布的特征Z的识别器D,之后最小化最大值的参数  
    
$$\underset{H}{min}\; \underset{D}{max}\;\; E_{p(Z|H)}[log D(Z)]+ E_{q_{\phi}(Z|X,S)}[log (1-D(Z))]$$  
  
  
<br>  

### 所以整体的目标函数为  
  
  
 $$\underset{H\;q_{\phi}}{min}\; \underset{D}{max}\;\;E_{q_{\phi}(Z|X,S)}[-log P(S|Z)]+ E_{p(Z|H)}[log D(Z)]+ E_{q_{\phi}(Z|X,S)}[log (1-D(Z))]$$  
   
     
<br>  
  
 利用GCN提取有效的空间特征。让后假设一个生成特征z的分布（从类别c中生成特征），让这个分布生成的Z接近提取的Z。完成训练后，可以利用假设的分布对特征进行分类。
