## 图卷积网络（GCN）   
  
  卷积网络：**的优势就是提取空间特征**
  
  CNN:处理的排列整齐的矩阵。（提取空间特征）  
    
 **但是还有一些非对齐的数据，如拓扑图。（如何提取空间特征呢?用到GCN）**  
   
  GCN: 中G就图论中拓扑图。（用GCN提取拓扑图形式数据的空间特征）  
  
![](img/2.png)  
    
 <br>  
 
 * 为什么研究GCN  
   
   1、CNN无法处理非对齐数据，在非对齐数据上（拓扑图数据结构）无法保持平移不变性。（无法使用相同的卷积核尺寸）  
     
   2、希望在（拓扑图数据结构）提取空间特征  
     
   3、广义上讲任何数据在赋范空间内都可以建立拓扑关联。（如谱聚类）  
     
<br>  

#### 提取拓扑图空间特征的两种方式  
  
  GCN的本质是用来提取拓扑图的空间特征。（目前有两种方式：1、vertex domain(spatial domain)顶点域（空间域 ） 2、spectral domain）  
    
 * 1、vertex domain(spatial domain)顶点域（空间域 ）  
   
   在拓扑图中，把每个顶点的相邻点找出来。两个问题1、按照什么条件找顶点的相邻点（neighbors），即如何确定接收域（receptive field） 2、确定了receptive field后，按照什么方式处理包含不同数目neighbors的特征。  
     
   这两个问题都方法都确定了之后,就可以提取特征了，大致方法为：  
 ![](img/1.png)  
   
   这种方法的缺点：1、每个顶点提取出来的相邻点不同，使得计算处理必须针对每个顶点。2、特征提取没有卷积好  
     
 <br>  
 
 * 2、spectral domain (GCN的基础)  
    
   **实现了拓扑图上的卷积操作**  
     
   处理graph问题时，使用到了拉普拉斯矩阵，GCN同样用到了拉普拉斯矩阵。    
     
<br>

### GCN公式  
  
  假设我们手头有一批图数据，其中有N个节点（node），每个节点都有自己的特征，我们设这些节点的特征组成一个N×D维的矩阵X，然后各个节点之间的关系也会形成一个N×N维的矩阵A，也称为邻接矩阵（adjacency matrix）。X和A便是我们模型的输入。  
    
  <br>  
  
  
GCN也是一个神经网络层，它的层与层之间的传播方式是：  
  
 $$Z^{(l+1)}=\sigma({\hat{D}}^{1/2}\hat{A}\hat{D}^{1/2})Z^{(l)}W^{(l)}$$  
   
 **其中Z是每层获得的特征。（输入层时是：X）**   
   
$\hat{A}=A+I$，I是单位矩阵   
  
$\hat{D}$是$\hat{A}$的度矩阵（degree matrix）  
  
σ是非线性激活函数  
  
W是每层的权重
