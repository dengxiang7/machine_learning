## 非线性可分SVM
非线性可分SVM,在特征空间中不存在线性超平面能将样本分类
![](../../img/Pattern_recognition/linear-spe-svm/non_linear_svm.PNG)  
  
    
不存在线性超平面使样本可分，但存在非线性超平面使样本可分：  
$a_1x_1+a_2x_1^2+a_3x_2+a_4x_2^2+a_5x_1x_2+a_6=0$  
  
    
那么对样本点做线性变换：  
$\phi(x)=\begin{bmatrix} x_1 \\ x_1^2  \\ x_2\\x_2^2\\x_1x_2\\ 1\end{bmatrix} $  
  
那么在这个变换的特征空间中，样本点线性可分：超平面为：$a^TX=0$  
  
<br>   
  
    
由于在特征空间中线性不可分，一定存在一个映射的高维空间，在这个高维空间中样本的映射是线性可分的，所以有以下推广。    
  
### 推广
样本点映射$\phi(x)$,映射空间的超平面：$y=W^T\phi(x)+b$  
  
那么原问题变为：  
$\qquad \quad \underset {w,b}{min} \; \frac{1}{2}\; ||w||^2$  
$\qquad \qquad s.t. y_i(W^T\phi(x) + b) \geq 1 \quad i=1,2\cdots,n$  
  
    
 <br>
 
 对偶问题:  
 $\qquad \quad \underset {\alpha}{max} -\frac{1}{2}\sum_{i=1}^n\sum_{j=1}^n\alpha_i\alpha_j y_i y_j \phi(x_i)^T\phi(x_j) +\sum_{i=1}^n\alpha_i$  
  $\qquad \quad \; s.t. \alpha \geq 0$  
  $\qquad \quad  \quad \sum_{i=1}^n\alpha_iy_i = 0$  
    
      
虽然映射到高维空间后线性可分，但却有两个问题：  
1.如何确定$\phi(x)$  
2.如何计算$\phi(x)^T\phi(x)$   
  
    
解决方法：核函数：$k(,.,) \quad\Rightarrow \;\; 查看数学笔记$
