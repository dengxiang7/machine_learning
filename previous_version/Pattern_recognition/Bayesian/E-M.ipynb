### E-M算法估计参数
#### 最大似然估计
$$\theta^*=\underset{\theta}{arg\;max\;}ln(P([x_1,z_1],\cdots,[x_n.z_n]|\theta))$$  
  
    
求最大似然估计时：若存在参数缺失，或者隐变量时，该怎么求最大的参数呢？ (这时给我们数据集，根据数据集我们知道的有 特征X(已知)，  
$\qquad\qquad\qquad\;\;\;$隐变量Z（未知），当前的参数$\theta$(中间概念)，最大化参数$\theta$(要求的))   
  
    
<br>    
  
例如：如果z是隐变量，如何求最大的参数。  
  
E步：在当前的X和$\theta^t$；估计Z的未知数为$\theta^{t+1}$的概率分布；并求出含$\theta^{t+1}$的Z的期望。  
  
    
$\qquad$Z的未知数为$\theta^t$的概率分布:$\;\; P(Z|X,\theta^t)$  
  
$\qquad$Z概率分布的期望:$\;\; E_z[P(Z|X,\theta^t)]$  
  
    
      
M步：求是期望$\;\; E_z[P(Z|X,\theta^t)]$最大的$\theta$    
    
<br>  

<br>  


### E-M 算法原理  
  
  令$X$表示已观测变量集，$Z$表示隐变量集，$\theta$表示模型参数。  
    
  我们的目标是：对$\theta$做极大化似然估计。(求使（X，Z）出现概率最大的$\theta$) $\;\; \underset{\theta}{arg\;max}\;\;ln P(X,Z|\theta)$  
    
  但是现在Z是隐变量，也就是Z是不知道的。  
    
  **在这种情况下：可以取遍Z,对Z求期望，最大化已观测数据的边际似然。$\;\; \underset{\theta}{arg\;max}\;\;ln \;\sum_{z}\; P(X,Z|\theta)$**  
    
  * E-M原理：  
     E步：根据已知$\theta^t$的值，推断出Z的取值，与分布。  
     M步：在得到Z值和Z的分布之后，我们就有（X，Z）都变得已知了。此时我们可以方便的重新对$\theta$做似然估计。$\;\; \underset{\theta}{arg\;max}\;\;ln \;\sum_{z}\; P(X,Z|\theta)$  
       
<br>  

* 流程：  
  E、基于$\theta^t$的值推断隐变量$Z$的期望，记为$Z^t$  
  M、基于已观测变量$X$和$Z^t$,对参数$\theta$做极大似然估计，记为$\theta^{t+1}$
  
<br>    

* 更近一步，基于$\theta^t$的值计算$Z$的概率分布$P(Z|X,\theta^t)$  
  
  E步、 以当前参数$\theta^t$推断隐变量$Z$的概率分布$P(Z|X,\theta^t)$ **这步$\theta^t$是具体值**。得到Z的分布之后，计算$\theta$在（X,Z）下关于Z分布的期望$\qquad\quad Q(\theta|\theta^t)=E_{Z|X,\theta^t}L(\theta|X,Z)$**这步$\theta$是未知参数**。  
  M步、寻找参数最大化期望似然，即$\theta^{t+1}=\; \underset{\theta}{arg\;max}\; Q（\theta|\theta^t）$
  
<br> 
  
<br>  

      
### E-M示例（GMM高斯混合模型）
高斯混合模型：将一个事物分解为若干的基于高斯概率密度函数（正态分布曲线）形成的模型。 为什么是“高斯混合模型”，而不是别的混合模型，因为从中心  
$\qquad\qquad\quad$极限定理知，只要K足够大，模型足够复杂，样本量足够多，每一块小区域就可以用高斯分布描述。而且高斯函数具有良好的计算性能，  
  
 $$p(x|\theta)=\sum_{i=1}^n\alpha_iN(x|u_i,\Sigma_i)$$  
   
   $$\sum_{i=1}^n\alpha_i=1$$   
     
       
 <br>  
   
用E-M算法估计高斯混合模型的参数（把$\alpha_i$看作隐变量）：  
  
  <br>
  
  E步（求$\alpha_i$的期望）：
  $$E_{\alpha}\;[lnp(X,\alpha|\theta)]$$  
    
     
     
  $$E_{\alpha}\;[\sum_{i=1}^n ln\;p(x_i,\alpha_i|\theta)]$$   
    
      
   $$=\sum_{i=1}^n E_{\alpha}ln\;p(x_i,\alpha_i|\theta)$$    
     
   $$=\sum_{i=1}^n \sum_{l=1}^k p(\alpha_i=l|x_i,\theta)\;\;ln\;p(x_i,\alpha_i=l|\theta)$$   
     
       
  <br>       
   
   其中$\alpha_i$在$x_i和\theta^t$下的概率分布：
   $$p(\alpha_i=l|x_i,\theta)=\frac{p(\alpha_i=l,x_i,\theta)}{p(x_i,\theta)}=\frac{N(x_i|U_l^t,\Sigma_l^t)*\alpha_i=l}{\sum_{l=1}^k N(x_i|U_l^t,\Sigma_l^t)*\alpha_i=l}\;\;\;\;@\gamma_{nl}^t$$  
     
       
  其中$\alpha_i$在$x_i和\theta^t$下的取值：  
  $$ln\;p(x_i,\alpha_i=l|\theta)\;=\;ln\;p(\alpha_i=l|x_i,\theta)p(x_i|\theta)\;=ln\;(\;N(x_i|U_l^t,\Sigma_l^t)*\alpha_i=l\;)$$  
    
      
  所以期望是 
  $$Q(\theta)=E_{\alpha}\;[lnp(X,\alpha|\theta)]=\sum_{i=1}^n \sum_{l=1}^k\;\;\gamma_{nl}^t\;*\;ln\;(\;N(x_i|U_l^t,\Sigma_l^t)*\alpha_i=l\;)$$  
    
  <br>  
  
M步：求使期望最大参数  
  
    
 $$\theta^{t+1}=\underset{\theta}{arg\;max\;}Q(\theta^t)=\underset{\theta}{arg\;max\;}\sum_{i=1}^n \sum_{l=1}^k\;\;\gamma_{nl}^t\;*\;ln\;(\;N(x_i|U_l^t,\Sigma_l^t)*\alpha_i=l\;)$$  
   
     
<br>  
  
对此式求导解出最值：  
  
  $$\frac{N(x_i|U_l^t,\Sigma_l^t)*\alpha_i=l}{\sum_{l=1}^k N(x_i|U_l^t,\Sigma_l^t)*\alpha_i=l}\;\;\;\;@\gamma_{nl}^t$$  
    
  <br>
  
  $$U^{t+1}=\frac{\sum_{i=1}^n\gamma_{nl}^tx_i}{\sum_{i=1}^n\gamma_{nl}^t}$$  
    
      
   <br>  
   
   $$\Sigma^{t+1}=\frac{\sum_{i=1}^n\gamma_{nl}^t(x_i-U_l^{t+1})(x_i-U_l^{t+1})^T}{\sum_{i=1}^n\gamma_{nl}^t}$$  
     
   <br>  
   
   $$\alpha^{t+1}=\frac{\sum_{i=1}^n\gamma_{nl}^t}{N}$$  
     
       
   <br>  
   <br>  
   ##### 理解
   虽然根据高斯混合模型知道，事件可以由多个高斯分布组合而成；但是在无标签的样本的集合中我们并不知道样本属于哪个高斯分布；如果知道样本属于哪个高斯分布；我们可以直接根据极大似然函数求出各个高斯分布的均值和方差；$\;\;\;\;$虽然现在我们不知道样本属于哪个高斯分布，（假设事件由K个高斯分布组成）；但是我们可以知道单个样本x在这K个高斯分布下的真实概率分布；那么我们就可以从样本真实概率的极大似然函数，变成求真实分布期望的极大似然函数
         
         
   ### E-M过程
   1、准备聚成几簇就把K设为几  
     
   2、解参数的过程就是对E-M过程进行迭代，知道参数收敛  
     
   3、每个样本属于哪个高斯分布，就是挑选$\gamma_{nl}^t $的最大值 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mxnet import  nd
from scipy import stats
import math


def get_pdf(sample, mu, sigma):
    res = stats.multivariate_normal.pdf(sample,mu, sigma)
    return res

def em(data, k, mu, sigma, steps):
    mu_p=[]
    k_p=[]
    
    for step in range(steps):
        num_gau = len(k)  # 高斯分布个数
        num_data = data.shape[0]  # 数据个数
        gama = np.zeros((num_gau, num_data))
        for i in range(num_gau): 
            for j in range(num_data):
                gama[i][j] = k[i] * get_pdf(data[j], mu[i], sigma[i]) / sum([k[t] * get_pdf(data[j], mu[t], sigma[t]) for t in range(num_gau)])
       #M步
        # 更新mu 均值
        mu_p_1=[]
        for i in range(num_gau):
            mu[i] = np.dot(gama[i], data) / np.sum(gama[i])
            mu_p_1.append(mu[i])
        mu_p.append(mu_p_1)
       
        
        # 更新sigma 协方差阵
        for i in range(num_gau):
            cov_sum = np.zeros((2, 2))
            for j in range(num_data):
                cov_sum += (gama[i][j]*(np.dot((data[j] - mu[i]).reshape(-1, 1), (data[j] - mu[i]).reshape(1, -1))))
            sigma[i] = cov_sum / np.sum(gama[i])
        # 更新k 概率
        k_p_1=[]
        for i in range(num_gau):
            k[i] = np.sum(gama[i]) / num_data
            k_p_1.append(k[i])
        k_p.append(k_p_1)
 
        print('step: {}\tk:{}'.format(step + 1, k))
        print('step: {}\tu:{}'.format(step + 1, mu))
        print('step: {}\tz:{}'.format(step + 1, sigma))
    
    return k, mu, sigma, gama, mu_p, k_p

def gen_example(num_fea,num_exam):
     #定义真实变量，模型参数
    #均值 u 标准差 Z  权重 a
    #u=[[-0.40658,0.32248],[1.20354,-1.19686],[-0.44149,-0.45088]]
    #z=[[[0.03446,0.01299],[0.1299,0.03458]],[[0.02259,0.00761],[0.00761,0.02361]],[[0.07024,0.03731],[0.03731,0.06498]]]
    z = [[[1.0, 0], [0, 10.0]],[[10.0, 0], [0, 1.0]],[[3.0, 0], [0, 4.0]]]
    u=[[1.0, 10.0],[10.0, 1.0],[10.0, 10.0]]
    a=[0.2,0.3,0.5]
    
    #生成样本 
    input_features=num_fea #特征数
    input_examples=num_exam  #样本数
    train_examples=np.zeros((num_exam,2)) #生成的训练样本
    a,b,c=0,0,0
    for i in range(input_examples):
        pros=nd.array([0,0.2,0.3,0.5])
        ui=nd.random.multinomial(pros)  #  按照给定的概率分布 决定 用哪个高斯分布生成样本
        if ui==1:
            train_examples[i]=np.random.multivariate_normal(u[0], z[0],1)  #生成二维正态样本
            
        if ui==2:
            train_examples[i]=np.random.multivariate_normal(u[1], z[1],1)
        if ui==3:
            train_examples[i]=np.random.multivariate_normal(u[2], z[2],1)
  
    return train_examples

def main(): 
    train_examples=gen_example(2,5000)
    #print(train_examples)
    # 原始数据散点图
    plt.scatter(train_examples[:, 0],train_examples[:, 1], s=5, c='r')
    plt.title('Raw Data')
    plt.show()
    
    #初始化参数
    a_a=[0.3,0.35,0.35]
    u_u=[[-0.5, -0.5],
         [-0.4, 0.4],
         [1.2, -1.2]]
    z_z=[[[1, 0.5], [0.5, 1]],
         [[0.3, 0], [0, 0.3]],
         [[0.5, -0.1], [-0.1, 0.5]]]
    #mean1 = [1.0, 2.0]
#mean2 = [2.0, 1.0]
#mean3 = [1.0, 1.0]
#cov1 = [[1.0, 0], [0, 10.0]]
#cov2 = [[10.0, 0], [0, 1.0]]
#cov3 = [[3.0, 0], [0, 4.0]]
    
    k, mu, sigma, gama,u_p,k_p=em(train_examples,a_a,u_u,z_z,25)
  
    
main()
    
```
