### 模糊C均值   
  
 在K-means上，允许一个样本属于多了类别，引入隶属度：$u_{ij}^b$ 样本i属于类别j的程度，b是事先给定的。    
   
 模糊C均值，用于解决K-means中的野值问题。  
   
<br>  
  
#### 目标函数  
  
  $$\underset {m,u}{min} \sum_{j=1}^c \sum_{i=1}^n \;u_{ij}^b ||x_i-m_j||^2$$  
      
 
 $$ \;\; s.t. \sum_{j=1}^c u_{ij}=1   \;\;\; u_{ij} >0$$  
   
<br>  
  
#### 优化  
  
  使用拉格朗日乘子法(暂时忽略非等式约束)：  
    
 $$L(u，m，\lambda)=\sum_{j=1}^c \sum_{i=1}^n \;u_{ij}^b ||x_i-m_j||^2  + \sum_{i=1}^n \lambda_i \left( \sum_{j=1}^c u_{ij}-1 \right)$$  
   
<br>  
  
$$\frac{\partial L}{\partial m_j}=\sum_{i=1}^n u_{ij}^b 2(m_j-x_i) =0 \;\;\;\; \Rightarrow \;\; m_j=\frac{\sum_{i=1}^n u_{ij}^b x_i}{\sum_{i=1}^n u_{ij}^b}$$    
  
<br>  
  
$$\frac{\partial L}{\partial u_{ij}}=b u_{ij}^{b-1} ||x_i-m_j||^2 +\lambda_i=0 \;\;\; \Rightarrow \;\; u_{ij}=\left(\frac{-\lambda_i}{b||x_i-m_j||^2 }\right)^{\frac{1}{b-1}}$$  
   
<br>  
  
 $$\sum_{j=1}^c u_{ij}=1 \;\;\;\ \Rightarrow \;\; (-\lambda_i)^{\frac{1}{b-1}}=\frac{1}{\sum_{j=1}^c \left(\frac{-\lambda_i}{b||x_i-m_j||^2 }\right)^{\frac{1}{b-1}}}$$   
   
<br>   
  
非负约束自动满足      
  
$$u_{ij}=\frac{\left(\frac{1}{||x_i-m_j||^2} \right)^{\frac{1}{b-1}} }{\sum_{k=1}^c \left(\frac{1}{||x_i-m_k||^2} \right)^{\frac{1}{b-1}} }$$  
  
    
<br>  
  
<br>  
  
### 算法步骤  
  
 1、初始化c个聚类中心$m_i \quad i=1,2\cdots,c$  
   
2、根据当前聚类中心计算每个样本的隶属度函数    
  
  $$u_{ij}=\frac{\left(\frac{1}{||x_i-m_j||^2} \right)^{\frac{1}{b-1}} }{\sum_{k=1}^c \left(\frac{1}{||x_i-m_k||^2} \right)^{\frac{1}{b-1}} }$$  
  
3、根据每个样本的隶属度，重新计算样本的中心$m_i \quad i=1,2\cdots,c$   
   
   $$ m_j=\frac{\sum_{i=1}^n u_{ij}^b x_i}{\sum_{i=1}^n u_{ij}^b}$$  
  
重复2，3 直到没到聚类中心$m_i \quad i=1,2\cdots,c$ 没有变化    
  
    
<br>  
  
<br>  
  
### 分析    
  
  对于野值$x_i \qquad ||x_i-m_j||^2 \quad$距离极大$u_{ij}$接近0，故解决了野值的问题。  
    
<br>  
  
  $b \rightarrow 1$那么$ u_{ij}\rightarrow 0 \; or \; 1$模糊C均值和K均值等价   
    
      
<br>   
  
* 约束隶属度和为1，可能对野值敏感，野值总要祸害某一个聚类。  
    
    
* 放宽约束，$\sum_{j=1}^c\sum_{i=1}^n u_{ij}=n$
