## 密度峰值聚类（DPCA)

#### 基本思想  
  
  聚类中心被低密度区域的样本近邻所包围。  
    
  聚类中心远离其他密度较高的点。  
    
<br>  
  
* 优点：  
  
  1、只基于数据点之间的距离  
    
  2、能够处理非球形的数据分布      
  
<br>  
    
* 缺点：  
  
  有时候很难确定聚类簇的数目  
    
  和一般的密度算法一样：都无法应对区域内密度均匀，不同区域密度变化的情况  
  
<br>  
  
#### 基本术语  
  
 1、局部密度    
   
 $\qquad$$\rho_i=\underset{j}{\sum} \chi(d_{ij} -d_c)$  
   
 $\qquad$指示函数$\chi(x) \quad$  x<0时函数返回1,$\quad$ 这个式子表明数据点i的局部密度为：距离数据点i的距离不超过阈值$d_c$的数据点个数  
   
<br>  

2、与其他高密度峰值点的距离：  
    
$\qquad$对于密度非最大的点，测量这些点与比这些点密度更大的点中，与它距离最小的点的距离      
      
$\qquad$$\delta_i = \underset{j:\rho_j>\rho_i}{min} (d_{ij})$   
  
  
$\qquad$$\delta_i$为密度峰值i，与那些密度峰值比i大的所有密度峰值点的距离中，选择与i距离最小的密度峰值点j   
  
<br>  
  
 $\qquad$如果这个点是密度最大点，没有密度峰值比它更大的点了，那么$\delta_i$设置为。  
  
$\qquad$$\delta_i = \underset{j}{max} (d_{ij})$   
  
  
   
     
<br>     
  
<br>  
  
#### 聚类过程  
  
  计算出每个数据点 ：的密度$\rho$ ， 和每个点与峰值较高点中的最近距离$\delta$。那么聚类中心就是那些密度又高，距离又远的点。  
    
![](../../img/Pattern_recognition/clustering/DPCA1.PNG)      
  
<br>  
  
 聚类中心是那些密度又高，距离又远的点。也就是取横纵坐标乘积最大得前几个。 (这里可以看出就是1，10)   
   
<br>  
  
而那些距离大（$\delta$大），但是密度低的点（$\rho$ 小），就是野值，（这里是26，27，28）  
   
<br>     
  
<br>  
  
#### 找到聚类中心之后的工作    
  
* 确定样本归属于那个聚类中心  
  
  把剩余样本分配给最近的（密度较高的）聚类中心。  
   
  不需要迭代优化
