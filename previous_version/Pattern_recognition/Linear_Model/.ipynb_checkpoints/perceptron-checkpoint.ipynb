### ÂâçÁΩÆÁü•ËØÜ
Â¢ûÂπøÁâπÂæÅÂêëÈáèÂíåËµ†ÂπøÊùÉÂêëÈáèÔºö  
$\qquad\qquad\quad g(y)=\alpha^Ty$  
$\qquad\qquad\quad \alpha=[w_0,w1,{\cdots},w_d]^T$  
$\qquad\qquad\quad y=[1,y1,{\cdots},y_d]^T$  
  
Á∫øÊÄßÊ®°ÂûãÈóÆÈ¢òÔºöÁé∞ÊúâËÆ≠ÁªÉÊ†∑Êú¨ÈõÜÊúâ ùëÅ‰∏™Ê†∑Êú¨{$ ùíö_1 , ùíö_2 , ... , ùíö_ùëÅ $}ÔºåÂÖ∂‰∏≠‰∏Ä‰∫õÊ†áËÆ∞‰∏∫$ùúî_1$Á±ªÔºåÂè¶‰∏Ä‰∫õÊ†áËÆ∞‰∏∫$ùúî_2$Á±ª„ÄÇÁî®Ëøô‰∫õÊ†∑Êú¨Á°ÆÂÆö‰∏Ä‰∏™Á∫øÊÄßÂà§Âà´ÂàÜÁ±ªÂô®ÔºåÂç≥Á°ÆÂÆöÂ¢ûÂπøÊùÉÂêëÈáèùú∂„ÄÇ‰ΩøÂæóÔºö  
  
$\qquad\qquad\quad   \begin{cases} \forall y_i \in \omega_1  \qquad \alpha^Ty_i>0  \\ \forall y_i \in \omega_2  \qquad \alpha^Ty_i<0 \end{cases} $   
  
    
Âú®Ê±ÇÁ∫øÊÄßÂà§Âà´Âô®‰πãÂâçÊàë‰ª¨ÂÖàÊääÊ†∑Êú¨ÈõÜÂÅö‰∏Ä‰∏ãËßÑËåÉÂåñÂ§ÑÁêÜ(ÊääÊ†∑Êú¨ÂèòÊç¢Âà∞Âêå‰∏Ä‰æßÔºåÊ≥®ÊÑèËøôÈáåÂèñË¥üÊòØÂ¢ûÂπøÂèñË¥ü)Ôºö  
  
$\qquad\qquad\quad   y_i=\begin{cases} y_i & if \quad y_i \in \omega_1  \\ -y_i & if \quad y_i \in \omega_2  \end{cases} $  
  
<br>    

$\qquad\qquad\quad \Rightarrow \alpha^Ty_i>0 \quad i=1,2,{\cdots},N$  

<br>  

Á∫øÊÄßÂàÜÁ±ªÂô®Ë¶ÅÊ≠£Á°ÆÂàÜÁ±ªÂè™Ë¶ÅÁ°ÆÂÆöÂ•Ω$\alpha $ÊñπÂêëÂ∞±Ë°å‰∫ÜÔºåÈÇ£‰πà$\alpha$ÂèñÂÄºËåÉÂõ¥ÊòØÂ§öÂ∞ëÂë¢Ôºü$\alpha^Ty>0 \;ËØ¥Êòé\alpha ÂíåyÁöÑÂ§πËßí‰∏∫ÈîêËßíÔºåÁé∞Âú®yÈÉΩÂú®Âêå‰∏Ä‰æß‰∫ÜÔºåÈÇ£‰πàÊª°Ë∂≥ÊâÄÊúây‰Ωø\alpha^Ty>0 ÁöÑÈò¥ÂΩ±Âå∫Â∞±ÊòØ\alphaÁöÑÊñπÂêëËåÉÂõ¥$

### ÊÑüÁü•Âô® 
##### Âü∫Êú¨ÊÄùË∑Ø
Ê†πÊçÆ‰∏äÂõæÁ°ÆÂÆö$\alpha$ËåÉÂõ¥ÁöÑËøáÁ®ãÂèØ‰ª•ÁúãÂá∫ÊÑüÁü•Âô®ÁöÑÂü∫Êú¨ÊÄùÊÉ≥Ôºö  
1.‰ªªÊÑèÂàùÂßãÂåñ‰∏Ä‰∏™$\alpha$ÂêëÈáè  
2.Ê†πÊçÆÂàÜÁ±ªÁªìÊûú‰øÆÊ≠£ÊùÉÂêëÈáè$\alpha$ÔºåÁÑ∂ÂêéÂÜçÂàÜÁ±ªÂÜç‰øÆÊ≠£ÔºåÁõ¥Âà∞ÂàÜÁ±ªÂÆåÂÖ®Ê≠£Á°Æ  
  
<br>
 Ê≥®ÊÑèÔºöÁî±‰∫éÁ∫øÊÄßÊ®°ÂûãÁöÑÂâçÊèêÊòØÁ∫øÊÄßÂèØÂàÜÁöÑ„ÄÇ‰∏îÂØπ‰∫éÊÑüÁü•Âô®ÁöÑÂâçÊèêË¶ÅÊòØÂÆåÂÖ®Á∫øÊÄßÂèØÂàÜÁöÑÔºåÊâçËÉΩÊâßË°åÂÆåÔºå‰∏çÁÑ∂‰ºö‰∏ÄÁõ¥‰øÆÊ≠£ÊùÉÂêëÈáè$\alpha$ÔºåË¶Å‰∏çÂ∞±ËÆæÁΩÆËø≠‰ª£‰∏äÈôê

Â¶Ç‰ΩïÊ±Ç$\alpha$‰ªé‰∏äÈù¢ÂèØ‰ª•ÁúãÂá∫ÊÑüÁü•Âô®ÁöÑÁõÆÊ†áÊòØÊâæÂà∞$\alpha ‰ΩøÊâÄÊúâ\alpha^Ty>0$ÈÉΩÊàêÁ´ãÔºåÊâÄ‰ª•ÊÑüÁü•Âô®ÁöÑÂáÜÂàôÂáΩÊï∞ÊàñÁõÆÊ†áÂáΩÊï∞‰∏∫Ôºö  
  
  $\qquad\qquad\quad \underset {\alpha}{min} J_f(\alpha)=\underset{\alpha^Ty_i<0} {\sum} (-\alpha^Ty_i)$  
    
  ÂØπËøô‰∏™ÂºèÂ≠êÂèØ‰ª•‰ΩøÁî®Ê¢ØÂ∫¶‰∏ãÈôçÊ≥ïÊ±ÇÊúÄ‰ºòËß£Ôºö 
  
  $\qquad\qquad\quad \frac{\partial J_f}{\partial \alpha}=\underset{\alpha^Ty_i<0} {\sum} (-y_i)$  
    
  $\qquad\qquad\quad \alpha (t+1)=\alpha (t) - \eta \frac{\partial J_f}{\partial \alpha}$  
    
  $\qquad\qquad\quad \alpha (t+1)=\alpha (t) + \eta \underset{\alpha^Ty_i<0} {\sum} (y_i)$  
    
  $ÂØπ‰∫é‰∏äÈù¢ÁöÑÊ±ÇËß£\alphaÊúÄ‰ºòËß£ÁöÑËøáÁ®ãÔºåËã•‰∏çËÆæÁΩÆËø≠‰ª£Ê¨°Êï∞ÁöÑ‰∏äÈôêÔºåÂàôÂâçÊèêÊòØÊ†∑Êú¨ÂÆåÂÖ®Á∫øÊÄßÂèØÂàÜ,Â≠òÂú®\;\alpha^* \;‰Ωø\;\underset {\alpha^*}{min} 
  J_f(\alpha^*)=0  $ÔºåÂê¶‰æßËÆæÁΩÆËø≠‰ª£Ê¨°Êï∞ÁöÑ‰∏äÈôêËÉΩÊâæÂà∞Ëøë‰ººËß£‰ΩøÊ†∑Êú¨‰∏çÂÆåÂÖ®ÂàíÂàÜ  
    
    
  ÂØπ‰∫é‰∏äËø∞ÁâõÈ°øÊ≥ïÊ±ÇÊúÄ‰ºòËß£ÔºåÊ†πÊçÆ$\eta$ÂíåËÆ°ÁÆó$\alpha$Êó∂Ê†∑Êú¨ÁöÑ‰∏™Êï∞ÊÉÖÂÜµËøòÂàÜ‰∏∫Âõ∫ÂÆöÂ¢ûÈáèÔºåÂèØÂèòÂ¢ûÈáèÔºåÂçïÊ†∑Êú¨‰øÆÊ≠£ÔºåÊâπÈáèÊ†∑Êú¨‰øÆÊ≠£Ôºö  
    
   Âõ∫ÂÆöÂ¢ûÈáèÔºåÂèØÂèòÂ¢ûÈáèÔºö$\qquad \eta(t)=1 \quad \eta(t)=\frac{|{\alpha(t)}^Ty|}{{||y||}^2}$  
     
       
   ÂçïÊ†∑Êú¨Âõ∫ÂÆöÂ¢ûÈáè‰øÆÊîπÔºö  
     $\qquad$1.‰ªªÊÑèÂàùÂßãÊùÉÂêëÈáè$\alpha(0),Âπ∂ËÆæt=0$  
     $\qquad$2.ËÄÉÂØüÊ†∑Êú¨$y_i,Ëã•{\alpha(t)}^Ty_i<=0,Âàô\alpha(t+1)=\alpha(t)+y_i $Âê¶ÂàôÁªßÁª≠  
     $\qquad$3.ËÄÉÂØüÂè¶‰∏Ä‰∏™Ê†∑Êú¨ÔºåÈáçÂ§ç2.Áõ¥Ëá≥ÊâÄÊúâÊ†∑Êú¨ÈÉΩÊúâ${\alpha(t)}^Ty_i>0 \;\;\;Âç≥ÔºöùêΩ_f(ùú∂)=0$  
       
         
   ÊïÖÂÆöÂ¢ûÈáèÊâπÈáè‰øÆÊîπ$\eta=1$Ôºö  
     $\qquad$1.‰ªªÊÑèÂàùÂßãÊùÉÂêëÈáè$\alpha(0),Âπ∂ËÆæt=0$  
     $\qquad$2.ËÄÉÂØüÊâÄÊúâÊ†∑Êú¨ÔºåËÆ∞ÂΩïÈîôÂàÜÊ†∑Êú¨ÈõÜY_err  
     $\qquad$3.$\alpha(t+1)=\alpha(t)+\eta \underset{y_i<0} {\sum} (Y_err)$  
     $\qquad$4.ÈáçÂ§çÊ≠•È™§2Áõ¥Ëá≥ÊâÄÊúâÊ†∑Êú¨ÈÉΩÊúâ${\alpha(t)}^Ty_i>0 \;\;\;Âç≥ÔºöùêΩ_f(ùú∂)=0$ 
    

   

```python
from mxnet.gluon import data as gdata
import numpy as np
import struct   #Â§ÑÁêÜÊï∞ÊçÆÊµÅ
import sys
from time import time
sys.path.append("..")
from matplotlib import pyplot as plt


#ÁîüÊàê‰∏§Á±ªÊ≠£ÊÄÅÊ†∑Êú¨
def genexam():
    z = [[[1.0, 0], [0, 10.0]],[[10.0, 0], [0, 1.0]]]
    u=[[1.0, 12],[12, 1.0]]
    a=[0.4,0.6]
    x1=[]
    x2=[]
    for i in range(1000):
        ui=np.random.multinomial(1,a,1)  #  ÊåâÁÖßÁªôÂÆöÁöÑÊ¶ÇÁéáÂàÜÂ∏É ÂÜ≥ÂÆö Áî®Âì™‰∏™È´òÊñØÂàÜÂ∏ÉÁîüÊàêÊ†∑Êú¨
        ui=ui[0]
        if ui[0]==1:
            x1.append((np.random.multivariate_normal(u[0], z[0],1))[0]) #ÁîüÊàê‰∫åÁª¥Ê≠£ÊÄÅÊ†∑Êú¨  
        if ui[1]==1:
            x2.append((np.random.multivariate_normal(u[1], z[1],1))[0])
    x1=(np.array(x1).T)
    x2=(np.array(x2).T)
    x1=np.row_stack((x1,np.ones((1,x1.shape[1]))))  #Ê∑ªÂä†‰∏ÄË°å‰∏∫1ÁöÑÊï∞ÊçÆ
    x2=np.row_stack((x2,np.ones((1,x2.shape[1]))))
    x3=np.hstack((x1,-x2)).T
    return x1,x2,x3


#ÁîªÂõæ
def draw1(xx1,xx2):
    plt.scatter(xx1[:, 0],xx1[:, 1], s=5, c='r')
    plt.scatter(xx2[:, 0],xx2[:, 1], s=5, c='b')
    plt.show()

def draw2(xx1,xx2,w_T,b_T1):
    w_T=np.array(w_T)
    fig = plt.figure(num=4, figsize=(7, 7),dpi=80) 
    ax1 = fig.add_subplot(1,1, 1)
    ax1.scatter(xx1[:, 0],xx1[:, 1], s=5, c='r')
    ax1.scatter(xx2[:, 0],xx2[:, 1], s=5, c='b')
    x = np.linspace(-5, 5,11)
    y = (-b_T1[0]*x-b_T1[2])/b_T1[1]
    xx = np.linspace(-5, 5,11)
    yy = (-w_T[0]*x-w_T[2])/w_T[1]
    ax1.plot(x, y)
    ax1.plot(xx, yy, c='r')
    plt.show()
 

 #ÂçïÊ†∑Êú¨Êõ¥Êñ∞W
def updateW(w_1,y,flag,num1):
    a=w_1.reshape(1,3)
    b=y.reshape(3,1)
    c=np.dot(a[0],b)
    if c[0]<=0:
        num1+=1
        flag=False  #ÊúâÈîôÁªßÁª≠Âæ™ÁéØ Âπ∂Êõ¥Êñ∞w
        w_1=w_1.reshape(3,1)+y.reshape(3,1)
    return w_1,flag,num1


#ÊâπÈáèÊ†∑Êú¨Êõ¥Êñ∞W
def batchupdateW(w_2,y):
    w_2=w_2.reshape(1,3)
    err=[]
    flag=True
    num2=0
    #Âæ™ÁéØ Áõ¥Âà∞ Ê≤°ÊúâÈîôËØØÂàÜÁ±ª
    while flag:
        #Âà§Êñ≠Ê†∑Êú¨ÊòØÂê¶ÂàÜÈîôÔºåÂØπÈîôËØØÁöÑÂä†ÂÖ•err
        for i in range(y.shape[1]):
            b=(y[:,i]).reshape(3,1)
            c=np.dot(w_2[0],b)
            if c[0]<=0:
                err.append(((y[:,i]).reshape(3,1)))
        err=np.array(err)
        #Êõ¥Êñ∞W
        for i in err:
            num2+=1
            w_2=w_2.reshape(3,1)+i
        #ÊâÄÊúâÊ†∑Êú¨ÈÉΩÊ≠£Á°ÆÂàÜÁ±ªÔºåË∑≥Âá∫Âæ™ÁéØ
        if len(err) == 0:
            flag=False
        else :
            w_2=w_2.reshape(1,3)
            #errÊ∏ÖÁ©∫
            err=[]  
        
    return w_2.reshape(3,1),num2


x=[[1,2,-3,-4],[4,3,-2,-1],[1,1,-1,-1]]
w=[[0],[0],[0]]
w=np.array(w)
x=np.array(x)


#ÂçïÊ†∑Êú¨Êõ¥Êñ∞
def singupdate(w,x):
    flage=True
    #ÂØπÊâÄÊúâÊ†∑Êú¨Êõ¥Êñ∞ÔºåÁõ¥Âà∞ÊâÄÊúâÊ†∑Êú¨Ê≠£Á°ÆÂàÜÁ±ª
    num1=0
    while flage:
        for i in range(x.shape[1]):
            a=w.reshape(1,3)
            b=(x[:,i]).reshape(3,1)
            c=np.dot(a[0],b)
            if c[0]<=0:
                num1+=1
                flag=False  #ÊúâÈîôÁªßÁª≠Âæ™ÁéØ Âπ∂Êõ¥Êñ∞w
                w=w.reshape(3,1)+(x[:,i]).reshape(3,1)
        flage=bool(1-flage)   #ÈÉΩÊ≠£Á°Æ flage ‰∏∫falseË∑≥Âá∫Âæ™ÁéØÔºå ËøòÊúâÊ≤°ÂàÜÂØπÁöÑÂàô flage ‰∏∫true ÁªßÁª≠Âæ™ÁéØ
    return w,num1
w1=singupdate(w,x)
w=[[0],[0],[0]]
w=np.array(w)
w=batchupdateW(w,x)



w=[[0],[0],[0]]
w=np.array(w)

#ÁîüÊàêÊ†∑Êú¨
xx1,xx2,xx3=genexam()
xx3=((np.mat(xx3)).T)
draw1(xx1.T,xx2.T)
start1=time()
w21,num1=singupdate(w,xx3)
end1=time()-start1
w=[[0],[0],[0]]
w=np.array(w)
start2=time()
w22,num2=batchupdateW(w,xx3)
end2=time()-start2


draw2(xx1.T,xx2.T,w21,w22)


print('ÂçïÊ†∑Êú¨ÔºåÂõ∫ÂÆöÂ¢ûÈáè')
print('Ëø≠‰ª£Ê¨°Êï∞Ôºö%d:'%(num1))
print('Ëä±Ë¥πÊó∂Èó¥Ôºö%f:'%(end1))
print('Âõ∫ÂÆöÂ¢ûÈáèÔºåÊâπÈáè‰øÆÊîπ')
print('Ëø≠‰ª£Ê¨°Êï∞Ôºö%d:'%(num2))
print('Ëä±Ë¥πÊó∂Èó¥Ôºö%f:'%(end2))
```
