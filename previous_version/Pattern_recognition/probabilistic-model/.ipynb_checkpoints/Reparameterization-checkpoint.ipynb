# 重参数  
  
  重参数是一种降低方差的技巧。**是处理$\color{red}{目标函数形式是，期望形式，的损失函数}的一种技巧。$**  
    
$$L_{\theta}(Z)= E_{z\text{~}p_{\theta}(Z)}[f(Z)]$$  
  
首先对这个期望形式目标函数的理解： **这个目标函数的变量是$\theta$ , 即：$L_{\theta}=p_{\theta}(Z_1)f(Z_1)+……+p_{\theta}(Z_m)f(Z_m)$**     
$\color{red}{假设现在现在函数f是固定的}$
   
常见的离散形式：  
  
$L_{\theta}= E_{z\text{~}p_{\theta}(Z)}[f(Z)]$  
  
$\quad\quad\;\; = \sum_{z}p_{\theta}(z)f(z)$  
  
连续形式：  
  
$L_{\theta}= E_{z\text{~}p_{\theta}(Z)}[f(Z)]$  
  
$\quad\quad\;\; = \int p_{\theta}(z)f(z) dz$    
  
**理解：f(Z)是目标函数。不同Z时，f(Z)有不同的输出。但是Z是服从参数为$\theta$的分布。不同的Z有不同的概率。所以经过f函数后，相当于对Z分布做了一个变换，变换为另一个分布。**

<br>      
  
<br>   
  
**假设现在我们要最小化$L_{\theta}$,调整z的分布的参数$\theta$,找到合适的值，让$L_{\theta}=p_{\theta}(Z_1)f(Z_1)+……+p_{\theta}(Z_m)f(Z_m)$ 期望最小。**    
    
**计算 $L_{\theta}$值得一种方法，就是通过，采样的方法。采样一个Z，有一个Z对应的概率P(Z),最后可以计算出$L_{\theta}$值。（MCMC马尔可夫链蒙特卡罗采样，最后趋于平稳分布。但是这种采样失去了参数$\theta$）**     
      
如果想使用梯度下降法，来通过对$\theta$求梯度，达到最小化$L_{\theta}$。那么我们就不能使用采样的方法,因为采样会失去参数$\theta$也就无法用梯度法了。  
  
**重参数技巧提供了一种变换方法。使我们在采样的同时，保留了参数$\theta$的梯度**   
  
<br>  
  
### 连续形式  
  
  $$L_{\theta} = \int p_{\theta}(z)f(z) dz$$   
    
 其中 $p_{\theta}(z)$具有显示的概率密度函数。在VAE中是正态分布。$N\text{~}N(z；\mu_{\theta},\sigma_{\theta}^2)$  
   
 要写出  $L_{\theta} $的表达式。一种式精确的积分。一种是采样。  
   
 保留$\theta$梯度的采样。  
   
 假设采样可以分为两步：1、从无参数分布的$q(\epsilon)$中采样一个$\epsilon$, 2、通过变换生成Z$\quad z=g_{\theta}(\epsilon)$ 。这个变换的参数是原本的参数$\theta$  
   
   $$L_{\theta}(Z)= E_{\epsilon\text{~}q(\epsilon)}[f(g_{\theta}(\epsilon))]$$  
     
**$\color{red}{对Z服从的分布，做一次变换。这时候被采样的分布就没有任何参数了，分布的参数都全部被转移到f内部了，}$**  
**$\color{red}{因此可以采样若干个点，当成普通的loss那样写下来了。}$**   
  
 <br>  
 
#### 例子  
  
  一个最简单的例子就是正态分布：对于正态分布来说，重参数为，从$N\text{~}N(z；\mu_{\theta},\sigma_{\theta}^2)$采样Z,变成从$N\text{~}N(\epsilon；0,1)$中采样一个$\epsilon$,然后计算$\epsilon \times \mu_{\theta} + \sigma_{\theta}$  
    
  所以$L_{\theta}(Z)= E_{z\text{~}N(z；\mu_{\theta},\sigma_{\theta}^2)}[f(Z)]$变为：$L_{\theta}(Z)= E_{\epsilon\text{~}N(\epsilon；0,1)}[f(\epsilon \times \mu_{\theta} + \sigma_{\theta})]$  
    
 **$\color{red}{现在通过无参数分布采样之后，再通过变换函数G,变换回原来的分布（注意要与原来显示的密度函数形式上一致）。}$**  
  **$\color{red}{这样即使用了采样的方法，又保留了分布的参数，可以使用梯度了。}$**

 从数学本质来看，重参数是一种积分变换，即原来是关于z积分，通过$z=g_θ(ε)$变换之后得到新的积分形式
