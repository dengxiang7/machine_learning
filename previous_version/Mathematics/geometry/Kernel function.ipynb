## 核函数
在低维空间线性不可分的点，可将样本点从原始空间映射到一个更高维的空间，使得样本在这个高维空间是线性可分的。

### 高维空间的计算    
虽然将样本映射到高维空间中是线性可分的，但还要解决样本在高维空间的计算问题：  
<br>
$\phi(x)^T\phi(x) \;\;$   
<br>
样本映射到高维空间点的内积运算；首先对于一个n维样本我们只知道一定存在一个高维空间使映射的样本线性可分，但我们却不一定知道这是一个什么的高维空间（映射函数），这个映射函数我们不好找；再则这个高维空间维数可能相当高。这都导致这个内积运算不好求。  
  
    
<br>
既然我们不知道这是一个什么样的映射函数，也不知道它是多少维。我们假设：在原始空间存在一个函数$K(x_i,x_j)$的计算结果等于高位空间$\phi(x)^T\phi(x)$内积：  
  
  $K(x_i,x_j)=\phi(x)^T\phi(x)$  
    
 <br>
 我们把$K(.,.)$称为核函数  
   
     
### 什么函数可以为核函数
虽然一个核函数$K(.,.)$可以绕开高维空间实现线性可分，但是什么的函数可以作为核函数也就是什么的函数满足$K(x_i,x_j)=\phi(x)^T\phi(x)$  
  
    
核函数： $\chi$为输入空间，$K(.,.)$是定义在$\chi \times \chi$上的对称函数，则$k$是核函数当且仅当对于任意数据$D=\{x_1,x_2,\cdots,x_n\}$,的核矩阵$K$是半正定的：  
  
<br>
 $$ K=\begin{pmatrix}
        k(x_1,x_1) & k(x_1,x_2) & k(x_1,x_3) & \cdots & k(x_1,x_n) \\
        k(x_2,x_1) & k(x_2,x_2)& k(x_2,x_3) & \cdots & k(x_2,x_n) \\
        \vdots & \vdots & \vdots & \ddots & \vdots \\
        k(x_m,x_1) &k(x_m,x_2) & k(x_m,x_3) & \cdots & k(x_m,x_n) \\
        \end{pmatrix}
$$  
  
<br>
这表明只要一个输入空间上的对称函数，它的核矩阵是半正定的那么这个函数就可以当作核函数来使用。当然这个核函数对应的高维空间的性能怎么样要验证。  
  
任何一个核函数都隐式的定义了一个称为‘再生核希尔伯特特征空间’。我们希望样本在特征空间内线性可分，因此特征空间的好坏对支持向量机的性能至关重要。  
  
    
      
<br>
在不知道特征映射函数时，我们并不知道什么样的核函数隐式定义的再生空间能够很好的线性可分。所以核函数的选择变成了最大的变数。若是核函数选择的不合适，则意味着将样本映射到一个不合适的特征空间。线性可分的性能可能不好。
