# 马尔可夫链蒙特卡罗采样（MCMC）  
    
## 蒙特卡罗法  
  
 **$\color{red}{\text{蒙特卡罗方法为：}}$**  
$$\theta =\int_a^bf(x)dx =\int_a^b p(x)\frac{f(x)}{p(x)}dx \approx \frac{1}{n} \sum_{i=0}^n\frac{f(x)}{p(x)}$$  
  
  可以看出来蒙特卡罗方法是用来求和或者求积分的。（蒙特卡罗方法不是用来采样的）（其中使用了独立同分布的大数定律）。  
    
  利用拒绝-接收，常见概率分布采样，我们可以使用蒙特卡罗方法。  
    
  **但是对于有些分布：  
  1、一些二维分布（x,y）,有时候只有条件分布P（X|Y),P(Y|X)，很难得到联合分布P（X，Y）一般形式，这时就无法用蒙特卡罗方法。  
  2、对于一些高维分布$P(x_1,……,x_n)$我们要找到适合的 k ,q(x)是很困难的。**  
    
  可以看出，要想将蒙特卡罗方法作为一个通用的采样模拟求和的方法，必须解决如何方便得到各种复杂概率分布的对应的采样样本集的问题。  
    
  蒙特卡罗方法不是采样方法，但是马尔可夫链蒙特卡罗方法是采样方法。

## 马尔可夫链    
   
 齐次马尔可夫模型：某一时刻的状态只依赖它的前一个状态。  
   
$$P(X_{t+1}|X_1X_2……X_{t-1}X_t)=P(X_{t+1}|X_t)$$  
  
**平稳分布：**  

$\{X_n,n=1,2,3,……\}$是一齐次马尔可夫链，其状态空间为E,一步转移概率矩阵为$P=(p_{ij})$。存在一确定的概率分布$\pi=\{\pi_i,i\in E\}$（$\pi$可以理解某一确定时刻 时，这个时刻取各个状态的概率）  
  
$\color{red}{\text{对这个确定的概率分布}\pi。\text{如果对任意时刻都满足：}\text{无论哪个时刻t,在这个时刻t时，取各个状态的概率是固定的。都是}\pi,\text{则}\pi\text{就是平稳分布}}$  
   
$$\pi^{t+1}_j=\sum_{i\in E} \pi^t_iP_{ij}$$   
  
由于每个时刻，取各个状态概率一样，都是平稳分布$\pi$,所以：  
  
  $$\pi_j=\sum_{i\in E} \pi_iP_{ij}$$
   
   

**平稳分布求法：**   
  
  $t_0$初始化一个随机分布$\pi^0=\{\pi_i^0,i\in E\}$ 。  
    
  根据边缘分布有$t_1$时刻，平稳分布$\quad\quad \pi_j^1=\sum_{i\in E} \pi_i^0P_{ij} \quad i,j \in E$ 。可得到$t_1$时刻平稳分布$\pi^1$  
    
  …………  
    
  …………  
    
  这样一直进行下去，最总某一时刻t之后，刻平稳分布$\pi^t \quad \pi^{t+1} ……$ 不再变化。可求得最总平稳分布。  
    
 ![](img/sampling_1.PNG)    
   
边缘分布$P（\pi^{t+1}_j）=\int P(\pi^t_i)P(ij)d\pi_i$  ;我们最总的目标是，求出每个时刻取各个状态概率都平稳的分布。虽然最总的平稳分布，是各个时刻都平稳的。但是在求平稳分布时。可以利用上诉公式，从一个随机的分布，传递，最总求出平稳分布。   
  
$\pi=\pi P$的解。就是平稳分布。    
  
<br>      
      
## 基于马尔可夫链采样  
  
  假设n轮之后平稳分布收敛，即
 $$\pi^n(x)=\pi^{n+1}(x)=…………$$  
  且对每个 $\pi（x）$有：
$$\pi^i（x）=\pi^{i-1}（x）P=\pi^{i-2}（x）P^2=……=\pi^{0}（x）P^i$$   
  
<br>  

$\pi^i（x）=\pi^{0}（x）P^i $:当i>n,表示平稳分布，可以由初始分布，经i次转移得到。这个公式可以用来获取平稳分布的采样。
    
<br>

现在开始采样：  
  
1、基于初始任意概率（比如高斯分布）$\pi^0(x)$采样，得到$t^0$的状态值$X^0$  
  
2、基于条件分布$P(X|X_0)$采样状态值$X_1$  
  
3、重复第2步，知道步数达到n次之后。每次在基于条件分布$P(X_{t+1}|X_t)$，采样状态值。（采的样本就属于平稳分布的样本了）  
  
<br>  
  
**总结马尔可夫链采样：**  
1、输入状态转移矩阵P,转移次数 阈值$n_1$,采样个数$n_2$  
  
2、从任意简单概率分布采样状态值$x_0$,  
  
3、$for \;\; t=0 \;\;to\;\;\;  t=n_1+n_2-1$从条件概率$P(X_{t+1}|X_t=X_0)$采样状态值$X_1$。最总我们只取后$n_2$个样本为我们需要的样本集$\{X_{n_1+1},X_{n_1+2},……,X_{n_1+n_2-1}\}$  
  
<br>    

<br> 

  
总结马尔可夫链采样存在的问题：  
  
我们可以得到我们需要采样样本的平稳分布所对应的马尔科夫链状态转移矩阵，那么我们就可以用马尔科夫链采样得到我们需要的样本集，进而进行蒙特卡罗模拟。   

**$\color{red}{意思是：把我们已知的（不好抽样的复杂分布）分布作为平稳分布。再给出这个平稳分布对应状态转移矩阵P。那么我们就可以从一个简单分布中抽样}$**  
**$\color{red}{再经过n次转移，为复杂平稳分布的抽样}$**
  
**但是一个重要的问题是，随意给定一个平稳分布π,如何得到它所对应的马尔科夫链状态转移矩阵P呢？**   
  
**幸运的是，MCMC采样通过迂回的方式解决了上面这个大问题。**

## 马尔可夫链蒙特卡罗采样（MCMC）  
  
  马尔科夫链中我们讲到给定一个概率平稳分布π, 很难直接找到对应的马尔科夫链状态转移矩阵P。而只要解决这个问题，我们就可以找到一种通用的概率分布采样方法，进而用于蒙特卡罗模拟。  
    
### 马尔科夫链的细致平稳条件  
  
  在解决从平稳分布π, 找到对应的马尔科夫链状态转移矩阵P之前，我们还需要先看看马尔科夫链的细致平稳条件。定义如下：  
    
  如果非周期马尔可夫链的状态转移矩阵为P,和概率分布π(x),对所有i，j都满足：  
    
 $$\pi(i)P(ij)=\pi(j)P(ji) \quad \text{细致平稳条件}$$  
   
 **则称概率分布$\pi(x)$是状态转移矩阵P(ij)的平稳分布**  
   
**也就是说，只要我们找到了可以使概率分布π(x)满足细致平稳分布的矩阵P即可。这给了我们寻找从平稳分布π, 找到对应的马尔科夫链状态转移矩阵P的新思路**    
  
**不过不幸的是，仅仅从细致平稳条件还是很难找到合适的矩阵P。比如我们的目标平稳分布是π(x),随机找一个马尔科夫链状态转移矩阵Q,它是很难满足细致平稳条件的，即**    
  
 $$\pi(i)Q(ij) \;\;\; \neq \;\;\; \pi(j)Q(ji)$$   
   
MCMC采样如何解决这个问题的。    
 
<br>  

### MCMC采样   
   
   一般情况下把我们的复杂分布，当平稳分布π(x)，我们再找一个状态转移矩阵P,一般来说随便找的P不满足细致平稳条件即：  
     
 $$\pi(i)Q(ij) \;\;\; \neq \;\;\; \pi(j)Q(ji)$$   
   
我们可以对上式做一个改造，使细致平稳条件成立。方法是引入一个$\alpha(i,j)$,使上式可以取等号，即：  
  
   $$\pi(i)Q(ij)\alpha(ij) \;\;\; = \;\;\; \pi(j)Q(ji)\alpha(ji)$$   
     
问题是什么样的α(i,j)可以使等式成立呢？其实很简单，只要满足下两式即可：  
  
 $$\alpha(ij)= \pi(j)Q(ji)$$  
 $$\alpha(ji)= \pi(i)Q(ij)\alpha(ij)$$  
   
 这样，就得到了分布π(x)对应的马尔科夫链状态转移矩阵P，满足：  
   
 $$P(ij)=Q(ij)\alpha(ij)  \qquad (\alpha(ij)\text{由上面两个公式得到})$$   
   
<br>  
  
也就是把我们的复杂分布，当平稳分布π(x)，再通过使细致平稳条件成立。可以找到这个复杂平稳分布的状态转移概率矩阵。此时我们就可以从一个简单分布中抽样，再经过n次转移，为平稳分布的采样。  
  
<br>   

α(i,j)我们有一般称之为接受率。取值在[0,1]之间，可以理解为一个概率值。即目标矩阵P可以通过任意一个马尔科夫链状态转移矩阵Q以一定的接受率获得

<br>
  
<br>  


#### MCMC采样总结  
  
1、输入我们(复杂)的平稳分布$\pi(x)$  ，输入任意选定的状态转移矩阵P,设定转移次数阈值$n_1$,设定采样样本个数$n_2$  

2、 已有$\pi(x)$和P,根据公式计算α(i,j)一般称为接受率。取值在[0,1]之间    
  
2、从任意简单概率分布中采样状态值$X_0$  
  
3、$for \;\; t=0 \;\;to\;\;\;  t=n_1+n_2-1$  
  
4、$\qquad\quad (a)$从条件转移概率$P(X^2|X^1=X_0)$ 采样一个状态值$X_1$  
  
5、$\qquad\quad (b)$从均匀分布U[0,1]采样一个$\mu$  
  
6、$\qquad\quad (c)$如果$\mu<\alpha(X_0，X_1)$  ，则接收转移$X_0 \rightarrow X_1$  
   
7、$\qquad\quad (d)$否则不接收转移$t=max(t-1,0)$   
  
8、最总得到样本集${X_{n_1+1},X_{n_1+2},……,X_{n_1+n_2-1}\}$    
  
<br>  

上面这个过程基本上就是MCMC采样的完整采样理论了，但是这个采样算法还是比较难在实际中应用，为什么呢？问题在上面第三步的c步骤，接受率这儿。由于 可能非常的小，比如0.1，导致我们大部分的采样值都被拒绝转移，采样效率很低。有可能我们采样了上百万次马尔可夫链还没有收敛，也就是上面这个 要非常非常的大，这让人难以接受，怎么办呢？这时就轮到我们的M-H采样出场了。  
  
<br>   
 
<br> 

#### M-H采样  
  
  解决采样率过低问题。  
    
$$\pi(i)Q(ij)\alpha(ij) \;\;\; = \;\;\; \pi(j)Q(ji)\alpha(ji)$$ 

$$\pi(i)Q(ij)\alpha(ij)b \;\;\; = \;\;\; \pi(j)Q(ji)\alpha(ji)b \qquad (\alpha(ji)b =1)$$  
  
$$\alpha(ij) \;\;\; = \;\;\; min\{\frac{\pi(j)Q(ji)\times 1}{\pi(i)Q(ij},1\} $$   
  
通过这个微小的改造，解决采样率第的问题。
