## 常用的优化算法
使用优化算法，不断迭代模型参数，去降低损失函数的值。  
  
在深度学习中，我们通常会定义一个模型函数，有了模型函数进一步定义损失函数，我们的优化算法就是试图让最小化损失函数。在优化中，这样的损失函数通常叫做优化的目标函数。  
  
深度学习中绝大多数目标函数都很复杂，因此，很多优化问题并不存在解析解，而需要使用基于数值方法的优化算法找到近似解，即数值解。（通常就是我们使用的优化算法）    
  
  
深度学习中优化问题面临的连个挑战。  
  
### 局部最小值  
局部最小值，但不是全局最小值。使用数值法去就全局最小值时。容易陷入局部最小值，而出不来的问题。  
  
  
### 鞍点  
局部最小值时，梯度接近或变成零。  
  
当解在鞍点附近时，也有 梯度接近或变成零。   
  
鞍点：（本质：一个不是局部最小点的驻点）。鞍点：微分方程中，沿着某一方向是稳定的，另一方向是不稳定的奇点，叫做鞍点。在泛函中，既不是极大值点，也不是极小值点的临界点，叫做鞍点。在矩阵中，一个数在所在行中是最大值，在所在列中是最小值，则被称为鞍点。在物理上，指在一个方向是极大值，另一个方向上是极小值的点。  
  
例如：$y=x^3$，中x=0的点。  
  
### 常用的数值优化算法  
  
* 1、梯度下降法：  
     每次沿着梯度的方向走一小步  
      
<br>      
      
* 2、随机梯度下降法（SGD）：  
    x为模型参数向量(不是样本值的向量)，n是样本个数，i为样本索引。 样本i的损失函数是$f_i(x)$，每一个样本的损失函数都是不同的，所以n个样本就有n个损失函数，（这n个损失函数的不同之处就体现在每一个损失函数的具体x值不同。；总的损失函数目标就是找到一个共同的x使所有的样本的损失函数和最小）。那么总的目标函数为：$f(x)=\frac{1}{n} \sum_{i=1}^n f_i(x) \quad$ 目标函数在x处的梯度 $\Delta f(x)=\frac{1}{n} \sum_{i=1}^n \Delta f_i(x) \quad$  
    
    正常的梯度下降法是：每一次迭代时，都是对所有样本总的损失函数$f(x)=\frac{1}{n} \sum_{i=1}^n f_i(x) \quad$求出的梯度$\Delta f(x)=\frac{1}{n} \sum_{i=1}^n \Delta f_i(x) \quad$  进行迭代 $x \leftarrow x - \eta \Delta f(x)$ 。直到达到最大迭代轮数后，逼近能使所有样本总损失函数值最小的点。  
     
     随机梯度下降法是：每一次迭代时，不再是求所有样本总损失函数的梯度进行更新。而是每一次随机选择一个样本的损失函数进行求梯度，对变量更新。$x \leftarrow x - \eta \Delta f_i(x)$。直到达到最大迭代轮数或者随机梯度逼近0。（这里随机一个样本的梯度，即$\Delta f_i(x)$是对$\Delta f(x)$的无偏估计）
