### 标准化 和归一化      
  
原文： https://blog.csdn.net/ruthywei/article/details/80980746  
  
#### 标准化  

* 数据的标准化（normalization）

    是将数据按比例缩放，使之落入一个小的特定区间。在某些比较和评价的指标处理中经常会用到，去除数据的单位限制，将其转化为无量纲的纯数值，便于不同单位或量级的指标能够进行比较和加权。  
      
<br>      
  
* 标准化最典型的方法：  
  
  就是数据的归一化处理，即将数据统一映射到\[0,1\]区间上。   
  
<br>    
  
* 标准化的其他方法：  
  
  直线型方法(如极值法、标准差法)、折线型方法(如三折线法)、曲线型方法(如半正态性分布)。  
  
不同的标准化方法，对系统的评价结果会产生不同的影响，然而不幸的是，在数据标准化方法的选择上，还没有通用的法则可以遵循。  
  
<br>  
  
#### 归一化  
  
* 归一化的目标：  
  
  1、把数变为（0，1）之间的小数  
    
    主要是为了数据处理方便提出来的，把数据映射到0～1范围之内处理，更加便捷快速。  
    
    <br>
    
  2、把有量纲表达式变为无量纲表达式。    
    
    归一化是一种简化计算的方式，即将有量纲的表达式，经过变换，化为无量纲的表达式，成为纯量。  
      
<br>  
  
* 归一化的好处：  
  
  1、提升模型的收敛速度（迭代优化的速度）  
    
     假如一个特征值的变化范围较大x1=0\~2000，一个特征值变化范围较小x2=0\~5；在对其进行优化时，求梯度是沿着等线垂直方向进行优化的，这样一个值变化较大，在另一个值上引起的波动较大（学习率，迭代的步长，使得原本两个值得范围就很大的值，跨步之后，偏离下一个等高线的垂直方向较远），就使得求梯度时，两个值互相牵引力大，梯度的波动大，梯度的优化路径就象一个之字形。迭代优化的速度很慢。反之，进行归一化后，两个值的互相牵引力就小的多，图更象一个圆形，优化的迭代速度会快很多。

 ![](../../img/math/numerical-process/normalization1.PNG)     
 
   2、 提升模型的精度  
      
   在涉及距离计算的时候效果明显。让各个维度的属性，对距离的贡献相差不要太大。
             
   3、防止深度网络中出现梯度爆炸  
     
   
<br>  
  
#### 常见的归一化的方法  
  
* min-max 标准化  
  
  对原始数据进行线性变化使结果落到\[0,1\]之间。min是数据样本的最小值，max是数据样本的最大值  
    
  $$x'=\frac{x'-min}{max-min}$$   
    
   缺点新样本来时要重新计算min和max  
     
       
<br>  
  
* z-score 标准化  
  
  标准差标准化($\mu$是样本均值，$\sigma$是样本标准差)：  
    
   $$x'=\frac{x-\mu}{\sigma}$$  
     
   要求原始数据的分布可以近似高斯分布，否则归一化的效果很糟糕。   
     
 <br>  
   
 * 非线性归一化(log函数转换)  
   
   该方法用于数据分化较大的场景，有些数值很大，有些数值很小，通过非线性函数将数据进行映射。（如log,正切等，视数据分布的情况决定使用什么非线性函数 ）  
     
   1、log对数函数转换法  
     
   <br>  
     
   2、atan反正切函数转换法  
     
   <br>  
     
   3、L2范数归一化方法
