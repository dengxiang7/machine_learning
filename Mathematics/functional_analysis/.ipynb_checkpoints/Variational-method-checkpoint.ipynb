# 变分法  
原文 ：https://zhuanlan.zhihu.com/p/41810184

### 基本概念
 
 变分法是**处理泛函的数学领域**，和处理函数的传统微积分相对。   
    

泛函：设{y}是给定的函数集，如果对于这个函数集中任一函数y(x) 恒有某个确定的数与之对应，记为П(y(x))，则П(y(x))是定义于集合{y(x)}上的一个泛函 **简言之，泛函就是函数的函数。**    
  

泛函的示例：1、两点之间的最短路径 2、最速降线问题 3、信息熵

   
<font color=red size=4>对泛函求极值的问题称为变分问题 </font>  使泛函取极值的函数称为变分问题的解，也称为极值函数。
    

传统的微积分中的一个常见的问题是找到一个 x 值使得 y(x) 取得最大值或者最小值。类似的，变分法中寻找一个函数 y(x)来最大化或者最小化泛函 П(y)  
   

**变分法可以用来说明两点之间的最短路径是一条直线或者最大熵分布是高斯分布。**  
 

- 算子是一个函数到另一个函数的映射，它是从向量空间到向量空间的映射
- 泛函是从**向量空间到数域的映射**
- 函数是从**数域到数域的映射**

### 引理  
 
* 引理1  

  如果函数$f(x)$在$[a,b]$内连续，$\eta(x)$是满足$\eta(a)=\eta(b)=0$的任意函数。如果$\int_a^b f(x)\eta(x)dx=0$成立。那么对于所有$x\in [a,b]$有$f(x)\equiv 0$成立 
    
  <br>  
  证明（反证法）：  
    
   设$f(x)$不恒为0。由于$\eta(x)$的任意性，可以设$\eta(x)=-f(x)(x-a)(x-b)$,x是a、b之间的变量。由于$f(x)$不恒为0，则有$\int_a^b -f^2(x)(x-a)(x-b)dx>0$成立。与条件相矛盾。所以假设不成立。则有$f(x)\equiv 0$成立。   
     
<br>  

* 引理2  
  
   如果函数$f(x)、g(x)$在$[a,b]$内连续，$\eta(x)、\xi(x)$是满足$\eta(a)=\eta(b)=0$且$\xi(a)=\xi(b)=0$的任意函数。如果$\int_a^b [f(x)\eta(x)+g(x)\xi(x)]dx=0$成立。那么对于所有$x\in [a,b]$有$f(x)\equiv g(x) \equiv0$成立  
       
  <br>  
  证明（反证法）：  
    
   设$f(x),g(x)$不恒为0。由于$\eta(x)、\xi(x)$的任意性，可以设$\eta(x)=-f(x)(x-a)(x-b)、\xi(x)=-g(x)(x-a)(x-b)$,x是a、b之间的变量。由于$f(x)、g(x)$不恒为0，则有$\int_a^b[ -[f^2(x)+g^2(x)](x-a)(x-b)]dx>0$成立。与条件相矛盾。所以假设不成立。则有$f(x)\equiv g(x) \equiv 0$成立。

### 函数的变分  
  
  对于任意定值$x\in[a,b]$，存在可取函数$y(x)$与另一可取函数$y_0(x)$。两函数在定值x处之间的差值$y(x)-y_0(x)$称为函数$y(x)$在函数$y_0(x)$处的$\color{red}{变分}$。记作$\delta y$。即函数的变分为$\delta y = y(x)-y_0(x)= \epsilon \eta (x) $  
  ![](img/functional_3.jpg)  
  
- 函数的变分 $\delta y$ 是两个不同的函数 y(x) 和 $y_0(x)$ 在自变量x固定时的差，这是函数发生了改变   
  
- 函数的增量 $\Delta y$ 是自变量x的增量使得函数 y(x) 产生的增量，函数依然是原来的函数。
    
<font color=red size=3> $\eta(x)$是在[a,b]上，两个函数的差函数。所以在x=a,和x=b的位置上有$\eta(a)=\eta(b)=0$  </font>
  
<br>

### 最简泛函的变分  
  
  设$F(x，y(x)，y\;'(x)\;)$是三个独立变量,且$x，y(x)，y\;'(x)\;$是在区间$[a,b]$的上的已知函数。且二阶连续可微，其中 y(x) 和 y'(x) 是x的未知函数，则泛函为：$J(y)=\int_a^b F(x，y(x)，y\;'(x)\;) dx \quad$这个泛函被称为最简的泛函，其中F是泛函的核。  
    
  在y=y(x)的一阶领域内，任取一曲线$y_1(x)$,则有变分$\delta y =y_1(x)-y(x) \;、\; \delta y\;' =y_1\;'(x)-y\;'(x)$  
    
  由泰勒展开式得最简泛函得增量为：   
    
  $\Delta J=J[y_1(x)] - J[y(x)] = J[\delta y(x)+y(x)] - J[y(x)] =\int_a^b F(x，y(x)+\delta y(x)，y\;'(x) + \delta y\;'(x)\;) dx\;-\; F(x，y(x)，y\;'(x) \;) dx \;\;= \int_a^b [F(x，y(x)+\delta y(x)，y\;'(x) + \delta y\;'(x)\;) \;-\; F(x，y(x)，y\;'(x) \;)]dx \;$  
  
  把这个式子得最后一项，叫做泛函的变分。记作$\delta J$

### Euler–Lagrange equation   
  
  回顾变分法理解1——泛函简介中的最速降线问题：  
  <img src="img/functional_4.jpg" width="250" hight="250">  
    
 设$\hat{y}(x)$是一个可取函数的类，即图中所有实线和虚线的集合。  
 在可取函数的类$\hat{y}(x)$中，存在一个二阶可微的**极值函数y(x)**，使得重物从O到A点的时间最短，即y(x)使得泛函$t=J(\hat{y})=\int_0^a\sqrt{\frac{1+(\hat{y}\;')^2}{2g\hat{y}}dx}$ 取得极值。  
 **用y(x)来表达所有的曲线函数**$\quad \hat{y}(x)=y(x)+\delta y=y(x)+\epsilon\eta(x)$  
 其中$\epsilon$为Lagrange引入的实数，$\eta(x)$是满足$\eta(a)=\eta(b)=0$的任意可谓的函数。  
 可以看出在$\epsilon = 0$时，有$\quad \hat{y}(x)=y(x)$,此时使得泛函$t=J(\hat{y})=\int_0^a\sqrt{\frac{1+(\hat{y}\;')^2}{2g\hat{y}}dx}$ 取得极值$J(y)$。  
 而泛函可以表示为：$\phi(\epsilon)=J[\hat{y}(x)]=J[y(x)+\epsilon\eta(x)]= \int_{x_0}^{x_1} F(x，y(x)+\epsilon\eta(x)，y\;'(x) + \epsilon\eta'(x) \;dx$  
 上式是**一个关于x的积分**，在完成积分后，剩下的只有$\epsilon$，所以是一个**一个关于$\epsilon$的函数**，   
 这个函数的特性是在$\epsilon=0$时，泛函取得极值$J(y)$。 也就是在$\epsilon=0$时，$\phi(\epsilon)$满足：$\frac{d\phi}{d\epsilon}|_{\epsilon=0}=0$  
 由求导与积分顺序的可交换性,$\frac{d\phi}{d\epsilon}$ 的形式可变形为（先求导后积分）：  
$\frac{d\phi}{d\epsilon}= \int_{x_0}^{x_1} F_y[(x，y(x)+\epsilon\eta(x)，y\;'(x) + \epsilon\eta'(x)]\eta(x) +  F_y'[(x，y(x)+\epsilon\eta(x)，y\;'(x) + \epsilon\eta'(x)]\eta'(x)\;dx$  
令$\epsilon=0$有：${d\epsilon}|_{\epsilon=0}=\int_{x_0}^{x_1}  [F_y\eta(x)+F_{y'}\eta'(x)]=0$  
利用分部积分法：  
$0=\int_{x_0}^{x_1}  [F_y\eta(x)+F_{y'}\eta'(x)]$  

$\quad  =\int_{x_0}^{x_1}  F_y\eta(x)dx+ \int_{x_0}^{x_1} F_{y'}\eta'(x)dx  $  
  
$\quad  =\int_{x_0}^{x_1}  F_y\eta(x)dx+ \int_{x_0}^{x_1} F_{y'}d\eta(x)  $   
  
$\quad  =\int_{x_0}^{x_1}  F_y\eta(x)dx+ F_{y'}\eta(x)|_{x_0}^{x_1} -\int_{x_0}^{x_1} \eta(x)d F_{y'} $
  
$\quad  =\int_{x_0}^{x_1}  F_y\eta(x)dx+ F_{y'}\eta(x)|_{x_0}^{x_1} -\int_{x_0}^{x_1} \eta(x)\left(\frac{d F_{y'}}{dx} \right)dx $  
  
$\quad  =\int_{x_0}^{x_1}  [F_y-\left(\frac{d F_{y'}}{dx} \right)]\eta(x)dx+ F_{y'}\eta(x)|_{x_0}^{x_1} $   
  
由于$\eta(x_0)=\eta(x_1)=0$,所以有 $ F_{y'}\eta(x)|_{x_0}^{x_1}=0$。进而推出有$ 0=\int_{x_0}^{x_1}  [F_y-\left(\frac{d F_{y'}}{dx} \right)]\eta(x)dx$   
  
再有引理一可以得到：$F_y-\left(\frac{d F_{y'}}{dx} \right)=0$恒成立。  
  
**这就是Euler–Lagrange equation**   
  
**$\eta(x)$ 实际上是加在y(x)上一个扰动，Euler–Lagrange equation的意义在于不管扰动是什么形式，如果泛函取得极值，都有 $F_y-\left(\frac{d F_{y'}}{dx} \right)=0$ 成立**

### 变分法的总结  
  
 最简泛函：为：$J(y)=\int_a^b F(x，y(x)，y\;'(x)\;) dx \quad$其中F是泛函的核。  
   
 **函数的变分：**$\delta y = y(x)-y_0(x)= \epsilon \eta (x)$   
   
 **泛函的变分：$\delta J =J(y_1)- J(y)$**
   
 **泛函的变分法：用泛函的变分来求泛函的极值是为变分法$\quad$由变分法可得：$F_y-\left(\frac{d F_{y'}}{dx} \right)=0$ 由此式解出的y，就是使泛函为极值的解y**   
   
 **（泛函变分法$F_y-\left(\frac{d F_{y'}}{dx} \right)=0$）的证明方法：  
 方法1、由泛函的变分$\delta J=0$可推到  
 方法2、（让泛函这个函数的变量）（为包含有变量函数的变分）（再直接对泛函求导，求出极值就可得到变分公式）$\phi(\epsilon)=J[\hat{y}(x)]=J[y(x)+\epsilon\eta(x)]$**  
   
 <br>  
   
<br>  

**变分法的公式：$F_y-\left(\frac{d F_{y'}}{dx} \right)=0$ 这个公式 称为E-L公式。解E-L公式就可以求出泛函的解**
