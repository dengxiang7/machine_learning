### 抽样分布
统计量：设$X_1,X_2,\cdots,X_n$是来自总体的一个样本，$g（X_1,X_2,\cdots,X_n）$是$X_1,X_2,\cdots,X_n$的函数，且g中没有其他的未知参数。那么$g（X_1,X_2,\cdots,X_n）$  
$\qquad\;\;\;\;\;$就是一统计量。  
  
$\qquad\;\;\;\;\;$例如$g（X_1,X_2,X_3）=5X_1+3X_2^2+7X_3 \quad$这里$g（X_1,X_2,X_3）$就是一统计量   
  
<br>  

样本$X_1,X_2,\cdots,X_n$都是随机变量；所以统计量g也是一个随机变量。而$x_1,x_2,\cdots,x_n$是样本的一次观察值；所以$g(x_1,x_2,\cdots,x_n)$是统计量g的一次观察值。  
    
 <br>   
 
    
* 常用统计量：  

    1、样本平均值： $\check{X}=\frac{1}{n}\sum_{i=1}^nX_i$  
      
    2、样本方差：$S^2=\frac{1}{n}\sum_{i=1}^n (X_i-\check{X})^2$  
      
    3、标准差：$S=\sqrt{S^2}$  
      
    4、样本K阶（原点）矩：$A_k=\frac{1}{n}\sum_{i=1}^nX_i^k$  
      
    5、样本K阶中心矩：$B_k=\frac{1}{n}\sum_{i=1}^n(X_i-\check{X})^k$  
      
        
 <br>  
 
###### 经验分布：  
  
  
我们观察到的有限个体$X_1,X_2,\cdots,X_n$：每一个个体记为样本  
  
概念上的所有样本的集合：记为总体X.  

  
现在用样本，定义一个与总体分布函数F(X)相近的统计量S $\Rightarrow$记为经验分布函数  
  
现在定义h(x)为样本个数;  

经验分布函数：$S(x)= \frac{1}{n}h(x) $  
  
设$x_1,x_2,\cdots,x_n$是总体X的n个观察样本。  
  
$S(x)= \frac{1}{n}h(x) =\begin{cases} 0, & x< x_1 \\ \frac{k}{n}, & x<x_k \\ 1, & x>x_n  \end{cases} $  
  
当样本个数n足够大时，经验分布与总体分布只有很小的差别。  
  
      
 <br>       
    
##### 抽样分布  
经验分布是一个统计量的分布（利用已得到的样本对这个统计量进行分布统计）；  
  
抽样分布也是一个统计量的分布（利用一个给定的统计量的分布，来抽取一定量的样本）。
