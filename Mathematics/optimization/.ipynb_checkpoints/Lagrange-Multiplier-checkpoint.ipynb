## 朗格朗日乘子法
可以将一个有n个变量于与k个约束条件的最优化问题转换为一个有n+k个变量方程组的解问题，引入了一个或一组新的未知数，新引入的未知数即为拉格朗日乘子。  
拉格朗日乘子法所得到的极点会包含原问题的所有极值点，但并不保证每个极值点都是原问题的极值点
  
### 前置知识
方向导数：函数在某点沿着某方向（向量）上的导数  
$\qquad\;\quad$如$u=u(x,y,z)在点p_0(x_0,y_0,z_0)处可微，则u=u(x,y,z)在点p_0处沿任一方向l_0的方向导数为：\frac{\partial u}{\partial l}={u_x}^\prime(p_0) \cos \alpha + {u_y}^\prime(p_0) \cos \beta +{u_z}^\prime(p_0) \cos \gamma$  
$\qquad\;\quad \cos \alpha \quad\cos \beta \quad \cos \gamma 为方向余弦$  
  
<img src="../../img/math/optimization/tidu1.png" width="500" high="500"/>
  
梯度：函数在给定点处沿不同的方向，其方向导数一般是不同的；$\color{red}{那么沿哪个方向其方向导数最大？最大值是多少？函数在点p_0沿哪个方向增加的速度最快？}$   

$\qquad u=u(x,y,z)在点p_0(x_0,y_0,z_0)处任意方向导数为：\frac{\partial u}{\partial l}={u_x}^\prime(p_0) \cos \alpha + {u_y}^\prime(p_0) \cos \beta +{u_z}^\prime(p_0) \cos \gamma$
$$\qquad\qquad \quad  l的方向\;\;=\;\;({u_x}^\prime(p_0)\cos \alpha,{u_y}^\prime(p_0)\cos \beta,{u_z}^\prime(p_0)\cos \gamma) $$

$\qquad 可以看出当方向l_0的方向余弦 \cos \gamma=1 \;\; \cos \alpha=1 \;\;\cos \beta=1$时方向导数最大，变化最快。所以$l_0方向为（{u_x}^\prime(p_0),{u_y}^\prime(p_0),{u_z}^\prime(p_0)）$   

$\qquad 定义梯度 grad \;u|p_0 =[{u_x}^\prime(p_0),{u_y}^\prime(p_0),{u_z}^\prime(p_0)] \quad 沿 grad \;u|p_0 变化最快，最大值为|grad \;u|p_0| $   
  
  
    
$\color{red}{梯度一个重要性质，是沿着变化速度最快的方向上升；\Rightarrow 等高线某点的梯度与该点的切线相切}$

<br>  
  
### 拉格朗日乘子法的由来  
描述：$min \;z=f(x,y)\;\;求z的最小值，即确定（x，y）;所以约束条件是关于x,y的约束条件；又如：u=f(x,y,z),则约束条件是关于x,y,z的$  
$如果min \;z=f(x,y)\;\;求z的最小值，而约束条件是g(x,y,z) =0则事实上约束条件上z对目标函数求最值不起作用，所以约束条件g(x,y,z)对x,y投影$  
$得到x,y的约束就行$
  
1.等式约束  
$\qquad \qquad \quad min \;z=f(x,y)$  
$\qquad \qquad \quad s.t. \; G(x,y,z)=0$

<img src="../../img/math/optimization/lagrange-Multiplier3.png" />
<img src="../../img/math/optimization/lagrange-Multiplier4.png"/>  
  
拉格朗日函数：  
$\qquad \qquad \quad L(x,\lambda)=f(x)+\lambda G(x)$  
$\qquad \qquad \quad 对个元素求导 \nabla_x L(x,\lambda)=\nabla f(x)+\lambda \nabla G(x)=0$  
$\qquad \qquad \quad\qquad \qquad \Rightarrow  \nabla f(x)=\lambda \nabla G(x)$  
$\qquad \qquad \quad\qquad \qquad \Rightarrow   G(x)=0$  
极值点满足的条件。  
  
2.不等式约束  
$\qquad \qquad \quad min \;z=f(x,y)$  
$\qquad \qquad \quad s.t. \; g(x,y)\leq 0$  
  
$如果最优点x^*在g(x)<0区域内部，则约束g(x) ≤0不起作用$  
<img src="../../img/math/optimization/lagrange-multiplier5.png" />  

<br>

$如果最优点x^*在g(x)=0边界上，则存在λ>0使得𝛻𝑓(𝒙^∗)+λ𝛻𝑔(𝒙^∗)=0$  
$λ为什么一定大于0？\quad 因为 𝛻𝑓(𝒙^∗) \;与\; 𝛻𝑔(𝒙^∗)反向$
$𝛻g(𝒙^∗)一定向外,因为外面g(x)>0,里面g(x)<0，而梯度方向是函数上升的方向；𝛻𝑓(𝒙^∗)一定向里,如果里面的𝑓(𝒙)比𝑓(𝒙^∗)小，那么就不会在边界取得最小值了$  

<br>

$综合以上两种情况在不等式约束下，要么g(x)<0， λ=0；则λg(x) =0 \;要么\;g(x)=0， λ>0，则λg(x) =0 故有 g(x) ≤ 0， λ≥0$  
  
  拉格朗日函数：  
$\qquad \qquad \quad L(x,\lambda)=f(x)+\lambda g(x)$  
$\qquad \qquad \quad 对个元素求导 \nabla_x L(x,\lambda)=\nabla f(x)+\lambda \nabla g(x)=0$  
$\qquad \qquad \quad\qquad \qquad \Rightarrow  \nabla f(x)=\lambda \nabla g(x)$  
$\qquad \qquad \quad\qquad \qquad \Rightarrow  g(x)\leq \;0$  
$\qquad \qquad \quad\qquad \qquad \Rightarrow  \lambda \geq \;0$  
$\qquad \qquad \quad\qquad \qquad \Rightarrow  \lambda g(x)= \;0$  
极值点满足的条件
  
    
      
<br>
<br>
所以对总体约束极值问题：  

$\qquad \qquad \quad min \;z=f(x,y)$  
$\qquad \qquad \quad s.t. \; g_i(x,y)\leq 0 \;\;i=1,2,{\cdots}$  
$\qquad \qquad \quad \quad \;\; h_i(x,y) = 0 \;\;i=1,2,{\cdots}$  
  
<br>    


转换为拉格朗日乘子问题：  
$\qquad \qquad \quad L(x,\lambda,\mu)=f(x)+\lambda g(x)+\mu h(x)\quad$   求这个无约束问题的极值问题。  
  
  
但是这个无约束的极值需要满足如下条件：  

$\qquad \qquad \quad
        \left.
        \begin{array}{l}
        \nabla_x L(x,\lambda,\mu)=0 \\
        \nabla_\mu L(x,\lambda,\mu)=0  \Rightarrow h_i(x,y) = 0\\
         g(x)\leq \;0  \\
        \lambda \geq \;0  \\
        \lambda g(x)= \;0 \\
        \end{array}
        \right\}
        = \;\;\;把这些条件称为KKT \;条件(\color{red}{最重要是解出\lambda，后可以解出使目标函数最小的x})
$  
  
<br>
<br>
<br>
      
### 原问题与对偶问题  
即使求出KKT条件，可以用来求最值，但是当我约束条件多的时候，用KKT条件来求最值任然是一个巨大的麻烦。
#### 前置知识
下界：给定一数集$E若存在m\in R 使得对 \forall x \in E 都有 x \geq m 则称m是集合E的一个下界\quad 多个下界$  
下确界：任意给定一数集$E$,我们称$E$的最大下界为$E$的下确界，记为$infE$,显然E中每个元素均大于或等于$infE\quad$ 多个下界，下确界是最大的下界  
##### 原问题
$\qquad \qquad \quad min \;z=f(x,y)$  
$\qquad \qquad \quad s.t. \; g_i(x,y)\leq 0 \;\;i=1,2,{\cdots}$  
$\qquad \qquad \quad \quad \;\; h_i(x,y) = 0 \;\;i=1,2,{\cdots}$    
  
##### 对偶问题  
  
$\qquad \qquad \quad
        \left.
        \begin{array}{l}
        g_i(x,y)\leq 0 \;\;i=1,2,{\cdots} \\
         h_i(x,y) = 0 \;\;i=1,2,{\cdots}\\
        \end{array}
        \right\}
        = \;\;\;把满足约束条件内的集合\;\;记为可行域 D
$  

<br>
  
$ L(x,\lambda,\mu)=f(x)+\sum_{i=1}^n\lambda_i g_i(x)+\sum_{j=1}^n\mu_j h_j(x)\quad$求这个拉格朗日在多个约束条件下满足KKT条件的解很麻烦，这样我们不求准确最小值，而求其近似值。    
  
    
定义对偶函数：  
$\qquad \qquad \quad \theta (\lambda ,\mu)= \underset {x \in R^d}{inf} L(x,\lambda,\mu)\quad 求L(x,\lambda,\mu)在KKT的解不好求，那么求 L(x,\lambda,\mu) 在R^d下的下确界来近似最值$
<br>      
$\qquad \qquad \quad 为什么可以近似：\underset {x \in R^d}{inf} L(x,\lambda,\mu)  \leq \underset {x \in D}{inf} L(x,\lambda,\mu)  = \underset{x \in D}{inf}[f(x)+\sum_{i=1}^n \underset{约束条件知 \leq 0}{\lambda_i g_i(x)}+\sum_{j=1}^n \underset{约束条件知 = 0}{\mu_j h_j(x)}]  \leq \underset{x \in D}{inf}f(x) \leq \underset{x \in D}{min \;f(x)}$  
<br>
$\qquad \qquad \quad \theta (\lambda ,\mu)= \underset{x \in R^d}{inf} L(x,\lambda,\mu)\qquad求解下确界\underset{x \in R^d}{inf} L(x,\lambda,\mu)的下确界是 \frac{\partial L}{ \partial x}=0\Rightarrow x表示\mu和\lambda 代入L就是具体的下确界值\theta (\lambda ,\mu)$
$ \qquad\quad\quad而下确界值\theta (\lambda ,\mu)是关于\lambda ,\mu的函数\Rightarrow所以求此函数的最大值就是\underset{x \in R^d}{inf} L的最大下确界；所以此时用这个最大下界来近似原问题最值。$  
      
<br>

$\qquad故对偶问题：$  
$\qquad \qquad \quad \underset {\lambda,\mu}{max} \;\theta (\lambda ,\mu)$  
$\qquad\qquad  \qquad \quad s.t. \;\mu\geq 0 \;$   
  
 <br> 

 $对偶问题的优势  $  
 $ \qquad \quad 可能比原问题好求解  $  
 $\qquad \quad无论原问题是否为凸规划，对偶问题都是凸规划  $  
  $\qquad \quad不过可惜，对偶问题的解未必是原问题的解  $  
 $对偶间隙  $  
 $\qquad \quad原问题与对偶问题解的差值$
