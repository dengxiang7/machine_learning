### 贝叶斯网  
贝叶斯置信网：借助有向无环图来描述属性之间的依赖关系，并使用条件分布来描述属性的联合概率分布。  
  
##### 结构  
有向无环图：$B=<G,\theta> \quad$网络结构G,其中节点对应属性，箭头代表属性之间的依赖$\quad \theta$顶来定理描述依赖关系。  
    
 <br>     
      
同父结构：儿子们条件独立  
$\quad x_3 \bot x_4|x_1 $  
$\quad p(x_1x_3x_4)=p(x_1)p(x_3|x_1)p(x_4|x_1) $  
![](../../img/Pattern_recognition/bayesian/bys2.PNG)  
  
    
<br>  

顺序结构：非后裔节点相互独立  
$\quad p(y\bot z|x)$  
$\quad p(yz|x)=p(z|x)p(y|x) $  
$\quad p(y|xz)=p(y|x) $  
$\quad p(x,y,z)=p(z)p(x|z)p(y|x)$
![](../../img/Pattern_recognition/bayesian/bys3.PNG)  
  
<br>  

V型结构：  
$\quad x_1 \bot x_2$  
$\quad p(x_1x_2)=p(x_1)p(x_2)$  
$\quad p(x_1x_2|x_4)\neq p(x_1|x_4)p(x_2|x_4)$ 
![](../../img/Pattern_recognition/bayesian/bys4.PNG)  
  
    
<br>

有向分离：分析变量之间条件独立性  
1、找出所有V型结构，在两个父节点之间加一个无向边  
2、所有有向边变无向边(道德图)  
3、如果变量集合z被去除则变量x和变量y分属两个分支，则变量x和变量y被z有向分离，变量x和变量y在条件Z下相互独立  
$\quad x_3 \bot x_4|x_1$  
$\quad x_3 \bot x_5|x_1$  
$\quad x_3 \bot x_2|x_1$  
$\quad x_5 \bot x_4|x_2$  
  
![](../../img/Pattern_recognition/bayesian/bys5.PNG)   
![](../../img/Pattern_recognition/bayesian/bys6.PNG)

###### 案例  
![](../../img/Pattern_recognition/bayesian/bys7.PNG)  

没使用贝叶斯网时计算p(x)概率：  
$p(x)=p(x_1)p(x_2|x_1)p(x_3|x_1x_2)p(x_4|x_1x_2x_3)p(x_5|x_1x_2x_3x_4)p(x_6|x_1x_2x_3x_4x_5)p(x_7|x_1x_2x_3x_4x_5x_6)$  
  
 <br>  
使用贝叶斯网：  
$p(x)=p(x_1)p(x_2)p(x_3)p(x_4|x_1x_2x_3)p(x_5|x_1x_3)p(x_6|x_3x_4)p(x_7|x_4x_5)$

### 如何寻找贝叶斯网  
评分搜索方式：  
1、定义一个评分函数；来判断贝叶斯网的依赖关系的好坏；越小说明贝叶斯与数据的契合度越高  
2、用评分函数寻找最优贝叶斯网  
  
    
<br>  

常用的评分函数：  

数据集$D=\{x_1,x_2,\cdots,x_n\}\quad$贝叶斯网$B=<G,\theta> \quad$  
那么在D上B的评分：  

$\qquad S(B|D)=f(\theta)|B|-LL(B|D)$  

$\qquad LL(B|D)=\sum_{i=1}^n log_2P_B(x_i)$  
    
 <br>   
    
$f(\theta)$ 描述参数所需的字节数，整个$f(\theta)|B|$表述网络复杂度；$LL(B|D)$为用网络B产生数据集D的概率  
  
   
  <br> 
  $p(x)=p(x_1)p(x_2|x_1)p(x_3|x_1x_2)p(x_4|x_1x_2x_3)p(x_5|x_1x_2x_3x_4)p(x_6|x_1x_2x_3x_4x_5)p(x_7|x_1x_2x_3x_4x_5x_6)$  
    
  不适用贝叶斯网时；是所有属性联合生成一个样本（一个样本呢所有属性联合）；  
    
      
  <br>  
  $p(x)=p(x_1)p(x_2)p(x_3)p(x_4|x_1x_2x_3)p(x_5|x_1x_3)p(x_6|x_3x_4)p(x_7|x_4x_5)$  
    
      
  使用贝叶斯网后是部分属性独立，部分属性依赖生成一个样本（不全部联合）   
    
      
 <br>  
 
 这样的话，基于两种方式的联合；对于同一个样本的产生概率必然不相同。所以需要度量贝叶斯网和样本集的契合程度。  
   
 <br>
 <br>
       
### 贝叶斯网的推断   
通过一些属性变量的观测值，来推测其他属性变量的取值是为推断，其中观测到的属性变量值视为证据。  
  
  ![](../../img/Pattern_recognition/bayesian/bys5.PNG)  
      
   <br>     
   依赖关系下取值推断概率（根据数据集可得）也是证据能用来推断：  
  ![](../../img/Pattern_recognition/bayesian/bys8.PNG)   
    
  <br>  
    
    
可以反向的推断：在$x_3=1,x_4=1,x_5=2$的情况下$x_1=1,x_2=1$的概率即$p(x_1=1,x_2=1|x_3=1,x_4=1,x_5=2)$：  

<br>  
  
$p(x_1=1,x_2=1 x_3=1,x_4=1,x_5=2)=p(x_1=1)p(x_2=1)p(x_3=1|x_1=1)p(x_4=1|x_1=1 ,x_2=1)p(x_5=2|x_2=1)=0.00288$  
  
$p(x_1=1,x_2=2|x_3=1,x_4=1,x_5=2)=p(x_1=1)p(x_2=2)p(x_3=1|x_1=1)p(x_4=1|x_1=1 ,x_2=2)p(x_5=2|x_2=2)=0.03528$  
  
$p(x_1=2,x_2=1|x_3=1,x_4=1,x_5=2)=p(x_1=2)p(x_2=1)p(x_3=1|x_1=2)p(x_4=1|x_1=2 ,x_2=1)p(x_5=2|x_2=1)=0.00252$  
  
$p(x_1=2,x_2=2|x_3=1,x_4=1,x_5=2=p(x_1=2)p(x_2=2)p(x_3=1|x_1=2)p(x_4=1|x_1=2 ,x_2=2)p(x_5=2|x_2=2)=0.04704)$  
  
  <br>  
    
$p(x_1=1,x_2=1|x_3=1,x_4=1,x_5=2)=\frac{p(x_1=1,x_2=1 x_3=1,x_4=1,x_5=2)}{p(x_3=1,x_4=1,x_5=2)}=\frac{0.00288}{0.00288+0.03528+0.00252+0.04704}=0.03283$
