# 推断  
  
 利用已知变量对未知变量进行推断。  
   
 例如：  在文本中,当打出文字为“我们的学校位”，利用这些已打出的文字，推断下一个文字为“置”“于”的概率。就是推断。即条件概率。  
   
 数学描述：  
   
  模型的变量集$x=\{x_1,x_2,……，x_n\}$，能分为两个不相交的子集$x_E \quad and \quad  x_F$。推断问题就是计算边际概率$P(x_F)$或者条件概率$P(x_F|x_E)$ ，最总计算出$P(x_E,x_F)$  
    
$$P(x_F|x_E)=\frac{P(x_E,x_F)}{p(x_E)}=\frac{P(x_E,x_F)}{\sum_{x_F}p(x_E,x_F)}$$  
  
推断即为计算条件概率：$P(x_F|x_E)$ 。  
  
**而这个条件概率记为：$P(x_{F1}\;\;\;\text{(x_F的一种可能)}|x_E)=\frac{x_{F1}\;\;\;\text{(x_F的一种可能)}}{\sum_{x_F}p(x_E,x_F) \;\;\;\text{(x_F的所有可能)}}$**  
  
**所以推断：的关键是计算所有可能的联合概率，$\sum_{x_F}p(x_E,x_F)$。而这个关键的问题在某些问题是不可求解的，或耗时极长的。现在要做的就是如何高效的求解所有可能联合概率，$\sum_{x_F}p(x_E,x_F)$**

### 近似推断算法  
  
  近似推断方法大致分为两类，第一类：采样，通过随机化方法完成。第二类：使用确定性近似完成近似推断。  
    
#### 采样法近似推断 ： MCMC 采样  
  
  采样的思想是：进行一些操作去逼近推断的概率分布即，边际分布$P(x_F)$或条件分布$P(x_F|x_E)$。这比直接计算精确值更为有效。  
    
  （个人想法：推断就是要去计算，边际分布$P(x_F)$或条件分布$P(x_F|x_E)$的值，来对未知变量$x_F$取什么值的概率最大，进行推断。除了精确的计算出概率的值外，我们可以对未知变量进行抽样，进行统计学分析，也可以近似的求出，边际分布$P(x_F)$或条件分布$P(x_F|x_E)$的值）  
    
 
 采样法的思路：  
   
   计算目标函数的期望(目标函数的输出就是我们要得到的值)。  
    
   假定我们的目标是计算函数$f(x)$在概率密度函数$p(x)$下的期望：  
     
   $E_p[f]=\int f(x)p(x)dx$  
     
   则可根据概率密度函数$p(x)$抽取一组样本$\{x_1,x_2,……，x_n\}$,然后计算$f(x)$的均值，$\hat{f}=\frac{1}{N}\sum_{i=1}^N f(x_i)$ 。我们可以利用这个均值来逼近目标函数的期望$E[f]$ 。这里基于大鼠定律，通过大量的采样是可以获得较高的精度的。  
     
   现在关键是如何采样：即**如何根据图模型所描述的概率(复杂分布)**去采样获取样本。  
    
   **概率图模型中最常用的采样方法是：MCMC**
