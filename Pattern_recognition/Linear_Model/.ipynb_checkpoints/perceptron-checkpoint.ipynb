## 感知器
### 前置知识
增广特征向量和赠广权向量：  
$\qquad\qquad\quad g(y)=\alpha^Ty$  
$\qquad\qquad\quad \alpha=[w_0,w1,{\cdots},w_d]^T$  
$\qquad\qquad\quad y=[1,y1,{\cdots},y_d]^T$  
  
线性模型问题：现有训练样本集有 𝑁个样本{$ 𝒚_1 , 𝒚_2 , ... , 𝒚_𝑁 $}，其中一些标记为$𝜔_1$类，另一些标记为$𝜔_2$类。用这些样本确定一个线性判别分类器，即确定增广权向量𝜶。使得：  
  
$\qquad\qquad\quad   \begin{cases} \forall y_i \in \omega_1  \qquad \alpha^Ty_i>0  \\ \forall y_i \in \omega_2  \qquad \alpha^Ty_i<0 \end{cases} $   
  
    
在求线性判别器之前我们先把样本集做一下规范化处理(把样本变换到同一侧，注意这里取负是增广取负)：  
  
$\qquad\qquad\quad   y_i=\begin{cases} y_i & if \quad y_i \in \omega_1  \\ -y_i & if \quad y_i \in \omega_2  \end{cases} $  
  
<br>    

$\qquad\qquad\quad \Rightarrow \alpha^Ty_i>0 \quad i=1,2,{\cdots},N$  

<br>  

线性分类器要正确分类只要确定好$\alpha $方向就行了，那么$\alpha$取值范围是多少呢？$\alpha^Ty>0 \;说明\alpha 和y的夹角为锐角，现在y都在同一侧了，那么满足所有y使\alpha^Ty>0 的阴影区就是\alpha的方向范围$

### 感知器 
##### 基本思路
根据上图确定$\alpha$范围的过程可以看出感知器的基本思想：  
1.任意初始化一个$\alpha$向量  
2.根据分类结果修正权向量$\alpha$，然后再分类再修正，直到分类完全正确  
  
<br>
 注意：由于线性模型的前提是线性可分的。且对于感知器的前提要是完全线性可分的，才能执行完，不然会一直修正权向量$\alpha$，要不就设置迭代上限

如何求$\alpha$从上面可以看出感知器的目标是找到$\alpha 使所有\alpha^Ty>0$都成立，所以感知器的准则函数或目标函数为：  
  
  $\qquad\qquad\quad \underset {\alpha}{min} J_f(\alpha)=\underset{\alpha^Ty_i<0} {\sum} (-\alpha^Ty_i)$  
    
  对这个式子可以使用梯度下降法求最优解： 
  
  $\qquad\qquad\quad \frac{\partial J_f}{\partial \alpha}=\underset{\alpha^Ty_i<0} {\sum} (-y_i)$  
    
  $\qquad\qquad\quad \alpha (t+1)=\alpha (t) - \eta \frac{\partial J_f}{\partial \alpha}$  
    
  $\qquad\qquad\quad \alpha (t+1)=\alpha (t) + \eta \underset{\alpha^Ty_i<0} {\sum} (y_i)$  
    
  $对于上面的求解\alpha最优解的过程，若不设置迭代次数的上限，则前提是样本完全线性可分,存在\;\alpha^* \;使\;\underset {\alpha^*}{min} 
  J_f(\alpha^*)=0  $，否侧设置迭代次数的上限能找到近似解使样本不完全划分  
    
    
  对于上述牛顿法求最优解，根据$\eta$和计算$\alpha$时样本的个数情况还分为固定增量，可变增量，单样本修正，批量样本修正：  
    
   固定增量，可变增量：$\qquad \eta(t)=1 \quad \eta(t)=\frac{|{\alpha(t)}^Ty|}{{||y||}^2}$  
     
       
   单样本固定增量修改：  
     $\qquad$1.任意初始权向量$\alpha(0),并设t=0$  
     $\qquad$2.考察样本$y_i,若{\alpha(t)}^Ty_i<=0,则\alpha(t+1)=\alpha(t)+y_i $否则继续  
     $\qquad$3.考察另一个样本，重复2.直至所有样本都有${\alpha(t)}^Ty_i>0 \;\;\;即：𝐽_f(𝜶)=0$  
       
         
   故定增量批量修改$\eta=1$：  
     $\qquad$1.任意初始权向量$\alpha(0),并设t=0$  
     $\qquad$2.考察所有样本，记录错分样本集Y_err  
     $\qquad$3.$\alpha(t+1)=\alpha(t)+\eta \underset{y_i<0} {\sum} (Y_err)$  
     $\qquad$4.重复步骤2直至所有样本都有${\alpha(t)}^Ty_i>0 \;\;\;即：𝐽_f(𝜶)=0$ 
    

   

```python
from mxnet.gluon import data as gdata
import numpy as np
import struct   #处理数据流
import sys
from time import time
sys.path.append("..")
from matplotlib import pyplot as plt


#生成两类正态样本
def genexam():
    z = [[[1.0, 0], [0, 10.0]],[[10.0, 0], [0, 1.0]]]
    u=[[1.0, 12],[12, 1.0]]
    a=[0.4,0.6]
    x1=[]
    x2=[]
    for i in range(1000):
        ui=np.random.multinomial(1,a,1)  #  按照给定的概率分布 决定 用哪个高斯分布生成样本
        ui=ui[0]
        if ui[0]==1:
            x1.append((np.random.multivariate_normal(u[0], z[0],1))[0]) #生成二维正态样本  
        if ui[1]==1:
            x2.append((np.random.multivariate_normal(u[1], z[1],1))[0])
    x1=(np.array(x1).T)
    x2=(np.array(x2).T)
    x1=np.row_stack((x1,np.ones((1,x1.shape[1]))))  #添加一行为1的数据
    x2=np.row_stack((x2,np.ones((1,x2.shape[1]))))
    x3=np.hstack((x1,-x2)).T
    return x1,x2,x3


#画图
def draw1(xx1,xx2):
    plt.scatter(xx1[:, 0],xx1[:, 1], s=5, c='r')
    plt.scatter(xx2[:, 0],xx2[:, 1], s=5, c='b')
    plt.show()

def draw2(xx1,xx2,w_T,b_T1):
    w_T=np.array(w_T)
    fig = plt.figure(num=4, figsize=(7, 7),dpi=80) 
    ax1 = fig.add_subplot(1,1, 1)
    ax1.scatter(xx1[:, 0],xx1[:, 1], s=5, c='r')
    ax1.scatter(xx2[:, 0],xx2[:, 1], s=5, c='b')
    x = np.linspace(-5, 5,11)
    y = (-b_T1[0]*x-b_T1[2])/b_T1[1]
    xx = np.linspace(-5, 5,11)
    yy = (-w_T[0]*x-w_T[2])/w_T[1]
    ax1.plot(x, y)
    ax1.plot(xx, yy, c='r')
    plt.show()
 

 #单样本更新W
def updateW(w_1,y,flag,num1):
    a=w_1.reshape(1,3)
    b=y.reshape(3,1)
    c=np.dot(a[0],b)
    if c[0]<=0:
        num1+=1
        flag=False  #有错继续循环 并更新w
        w_1=w_1.reshape(3,1)+y.reshape(3,1)
    return w_1,flag,num1


#批量样本更新W
def batchupdateW(w_2,y):
    w_2=w_2.reshape(1,3)
    err=[]
    flag=True
    num2=0
    #循环 直到 没有错误分类
    while flag:
        #判断样本是否分错，对错误的加入err
        for i in range(y.shape[1]):
            b=(y[:,i]).reshape(3,1)
            c=np.dot(w_2[0],b)
            if c[0]<=0:
                err.append(((y[:,i]).reshape(3,1)))
        err=np.array(err)
        #更新W
        for i in err:
            num2+=1
            w_2=w_2.reshape(3,1)+i
        #所有样本都正确分类，跳出循环
        if len(err) == 0:
            flag=False
        else :
            w_2=w_2.reshape(1,3)
            #err清空
            err=[]  
        
    return w_2.reshape(3,1),num2


x=[[1,2,-3,-4],[4,3,-2,-1],[1,1,-1,-1]]
w=[[0],[0],[0]]
w=np.array(w)
x=np.array(x)


#单样本更新
def singupdate(w,x):
    flage=True
    #对所有样本更新，直到所有样本正确分类
    num1=0
    while flage:
        for i in range(x.shape[1]):
            a=w.reshape(1,3)
            b=(x[:,i]).reshape(3,1)
            c=np.dot(a[0],b)
            if c[0]<=0:
                num1+=1
                flag=False  #有错继续循环 并更新w
                w=w.reshape(3,1)+(x[:,i]).reshape(3,1)
        flage=bool(1-flage)   #都正确 flage 为false跳出循环， 还有没分对的则 flage 为true 继续循环
    return w,num1
w1=singupdate(w,x)
w=[[0],[0],[0]]
w=np.array(w)
w=batchupdateW(w,x)



w=[[0],[0],[0]]
w=np.array(w)

#生成样本
xx1,xx2,xx3=genexam()
xx3=((np.mat(xx3)).T)
draw1(xx1.T,xx2.T)
start1=time()
w21,num1=singupdate(w,xx3)
end1=time()-start1
w=[[0],[0],[0]]
w=np.array(w)
start2=time()
w22,num2=batchupdateW(w,xx3)
end2=time()-start2


draw2(xx1.T,xx2.T,w21,w22)


print('单样本，固定增量')
print('迭代次数：%d:'%(num1))
print('花费时间：%f:'%(end1))
print('固定增量，批量修改')
print('迭代次数：%d:'%(num2))
print('花费时间：%f:'%(end2))
```
