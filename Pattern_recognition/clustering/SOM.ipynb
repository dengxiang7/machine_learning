### SOM自组织映射网络  
竞争型神经网络的典型代表，SOM网络：不同的区域对应不同的功能。对于不同类型的刺激，响应出现在不同的区域。  
  
SOM网络：  
1、神经元都在同一层上，呈规则排列，样本的特征向量的每一维通过权值全连接到网络的每一个节点上。   
  
2、神经元节点之间没有直接连接，然而在学习过程中相邻神经元之间有一定的相互影响，构成邻域相互作用  
  
3、神经元的计算功能就是对输入的样本给出响应  
  
4、一个节点对输入样本的响应强度就是该节点的权值向量与输入向量的匹配程度  
  
5、对于一个输入样本，神经元平面上所有的节点中响应最大的节点称为获胜节点

### SOM网络训练的过程
1、用小的随机数初始化权值向量，注意各节点的初始向量不能相等。重复2到4  
  
2、在$t$时刻按照给定的顺序或随机的给一个样本$x_t$  
  
3、计算神经元响应，找到响应最大(权值向量与输入向量的匹配程度最大)的神经元为获胜神经元$c$($t$时刻获胜神经元为$c$) ，(神经元平面为A，$t$时刻样本$x_t$  
$\quad$与$t$时刻$i$神经元$m_i^t$之间的距离为d)获胜神经元满足的条件是$d(x_t,c)=\underset{i \in A}{min}d(x_t,m_i^t)$  
  
4、权值竞争学习，对所有神经元做如下更新。  
$\quad  m_i^{(t+1)}=m_i^t\;+\;\alpha^t \;h_{ci}^t\;(x_t-m_i^t) $ $\quad（* \alpha^t为学习步长 0到1 之间 \;\;$  $\;h_{ci}^t\; 为$t时刻获胜节点$c$  与其它神经元节点$i$是否是近邻是近邻$\;h_{ci}^t=1 否则\;h_{ci}^t=0$
  
 
  
 ### 分析
1、 $\alpha^t 和 \;h_{ci}^t$初始值开始设置得较大，逐渐减小，邻域到算法终止时只包含自己  
  
2、没有明确的目标函数，有研究表明，如果邻域固定，则SOM等价于通过梯度下降法来最小化  
  
自组织特性:  
  
1、随着学习过程的进行，对于某个输入样本x ，对应的获胜节点i将趋于稳定，称为样本x的像。 x称为节点i的原像
  
2、一个样本只能有一个像，一个神经网络可以有多个原像，或者没有像  
  
    
###  SOM用途
1、拓扑保持的映射：SOM完成从原样本空间到二维平面上神经元网络的映射。在原空间中样本间的距离关系在只有有限个节点的平面网格上得到尽可能的保持  
  
 2、高度压缩的映射：原空间高密度区域的样本被“挤压”在少数节点上。   
   
### SOM局限性  
1、网络结构固定，不能动态改变  
  
2、网络训练时，有些神经元始终不能获胜，成为“死神经元”  
  
3、当输入数据较少时，训练的结果通常依赖于样本的输入顺序  
  
4、网络连接权的初始状态、算法中的参数选择对网络的收敛性能有较大影响  
  
  
###  SOM聚类
1、随着学习的不断进行，学习率不断减小， 邻域也将不断缩小，所有权向量将在输入向量空间相互分离 ，各自代表输入空间的一类模式，这完成了聚类  
  
2、一个神经元节点代表一个聚类中心，有多少个神经元就有多少个聚类  
  
3、如果想聚成较少个数的簇，可以再对神经元节点聚类
