### 聚类任务
无监督的学习任务:，也就是训练数据集中的样本都是无标记。我们的目标就是对无标记的样本集进行学习，希望能够学到样本集中的潜在规律  
$\qquad\qquad\qquad\quad$和内在性质，从而利用这些性质和规律去判别未见样本。  

    
 <br>
 
聚类任务：是无监督的学习任务中的代表。聚类任务的目标是把样本集划分为若干个不相交的子集，每个子集对应一个潜在的概念也是对应一个类别，  
$\qquad\qquad$把这个子集叫做簇。   
  
<br>  
  
聚类的类型：更具不同的学习策略，形成了原型聚类，密度聚类，层次聚类。  
  
<br>  
  
#### 聚类的两个基本问题： 
一、性能度量  
性能度量：也叫有效性指标，即聚类结果好坏性的评估。也用作聚类优化的目标。  
  
两大类：1、将聚类结果与某个参考模型进行比较，称为外部指标。  
$\qquad\quad$2、直接考察聚类结果而不利用任何的参考模型，称为内部指标。
  
 <br>  
    
 外部指标：  
 数据集$D=\{x_1,x_2,\cdots,x_n\}$ ,聚类结果：$C=\{c_1,c_2,\cdots,c_k\}$,真实聚类结果$C^*=\{c_1^*,c_2^*,\cdots,c_k^*\}$,且$\lambda,\lambda^*$分别为样本在聚类结果的标记，和在真实聚类中标记。  
   
   $$a=|SS|,\;\;SS=\{(x_i,x_j)|\lambda_i=\lambda_j^*\;\;\;\lambda_i=\lambda_j^*,\;\;i<j\} $$   
     
   $$a=|SD|,\;\;SD=\{(x_i,x_j)|\lambda_i=\lambda_j^*\;\;\;\lambda_i\neq\lambda_j^*,\;\;i<j\} $$   
     
   $$c=|DS|,\;\;DS=\{(x_i,x_j)|\lambda_i\neq\lambda_j^*\;\;\;\lambda_i=\lambda_j^*,\;\;i<j\} $$   
     
   $$c=|DD|,\;\;DD=\{(x_i,x_j)|\lambda_i\neq\lambda_j^*\;\;\;\lambda_i\neq \lambda_j^*,\;\;i<j\} $$  
     
       
         
  SS集合为：在聚类中相同簇，在真实聚类中也是相同簇的样本对       
    
      
1、Jaccard系数：$\frac{a}{a+b+c}$   

2、FM指数：$FMI=\sqrt{\frac{a}{a+b}*\frac{a}{a+c}}$  
  
3、Rand指数：$RI=\frac{2(a+b)}{n(n-1)}$   
  
这3个指标在0-1之间，且越大越好。    
  
 <br>   
      
 内部指标：  
 $avg(c)\;\;$对应簇C的样本平均距离  
   
$diam(c)\;\;$对应簇C的样本间最远距离  
  
$d_{min}(c_i,c_j)\;\;$对应簇$C_i与簇C_j$的最近样本之间的距离   
  
    
 $d_{cen}(c_i,c_j)\;\;$对应簇$C_i与簇C_j$的中心点的距离   
   
     
DBI指数：$DBI=\frac{1}{K}\sum_{i=1}^k \underset{i\neq j}{max}(\frac{avg(c_i)+avg(c_j)}{d_{cen}(c_i,c_j)})$   
  
    
Dunn指数：$DI=\underset{1\leq i\leq k}{min }\{\underset{j\neq i}{min\;}\;(\frac{d_{min}(c_i,c_j)}{\underset{1\leq l\leq k}{max\;\;\;}diam(c_l)})\}$
