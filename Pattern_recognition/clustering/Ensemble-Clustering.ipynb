### 集成聚类  
  
  集成聚类:也叫一致性聚类或者聚合聚类，旨在将多个现有的基本分区融合成一个完整的分区。  
    
<br>  

根据相似性的不同等级现有的工作可以大致分为两类：  
  
1、  第一类使用效用函数来度量一致性聚类和多个基本分区在分区级别上的相似性。这种方法通常通过最大化效用函数值来寻找最终的聚类结果。  
 
 
 $\qquad\quad$例如，提出了一个基于二次互信息的目标函数来进行一致性聚类，并进一步扩展了他们的工作，使用有限多项分布混合的EM算法来进行一致性聚类,  
$\qquad\quad$沿着这条思路，Wu等人提出了基于kmeans的一致性聚类(KCC)，并将一致性聚类转化为具有不同KCC效用函数的K-means聚类问题  
  
<br>

2、将基本分区的信息汇总到一个联合关联矩阵中，该矩阵一种计算方式是计算两个实例出现在同一个聚类中的次数。因此，该矩阵可以看作是基于实例层次的相似矩阵，在此基础上可以进行任何图划分算法进行最终聚类。  
  
   $\qquad\quad$ 例如，凝聚聚类，例如，谱聚类方法，（得到的部分的指示向量H）将其转化为加权的K-means问题，并将两种集成聚类方法联系起来。

<br>

#### 集成聚类  
  
  集成聚类的目标是：尽可能找到尽可能与现有的多个基本分区一致的单个一致性分区。集成聚类有助于生成健壮的分区、发现奇异的聚类、处理噪声、离群值和样本变化，以及集成来自多个分布式或不完整的数据源或属性的解决方案，这一点已得到广泛认可。  
    
 <br>
 
 #####  对于第二类集成聚类，如何进行聚类的集成
 
   
给出数据矩阵X的r个基本分区$H=\{H^1,H^2,\cdots,H^r\} \qquad H^i$基本分区的簇标号为$K_i$,目标是将多个现有的基本分区融合成一个完整的分区。  
     
基本分区的产生：不同参数的同一聚类算法，不同特征的同一聚类算法，甚至不同的聚类算法都可能产生基本的分区。   
   
 <br>  
     
 联合矩阵的生成：将基本分区的信息汇总到一个联合关联矩阵中，该矩阵计算两个实例出现在同一个聚类中的次数    
 
   
 联合矩阵： $S(x,y)=\frac{1}{r} \sum_{1}^r \delta (H^i(x),H^i(y))$  
    
 $\qquad\qquad\quad$ （共生矩阵）记录 样本对 共同出现的 基本分区的个数    
   
     
将集成聚类问题重新定义为一个经典的图划分问题，从而在相关矩阵S的基础上，推导出一些聚类规则或损失函数，从而得到最终的一致聚类  

<br>

  定理：虽然由于生成策略的不同，基本分区之间可能存在很大的差异，但随着基本分区个数r的增加，归一化协相关矩阵变得稳定
