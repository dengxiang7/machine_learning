   1 of k coding (类型特征转换 1 of k 哑编码)：将非数字型的特征值转换为数值型数据  
     
   假设变量的取值有k个，对这些值用1到k编码，则可以用维度为k的向量来表示一个变量的值：  
     
   假设变量字母：取值为 ：A，B，C  ，对取值编码为：1，2，3   
     
   取值A:用k维向量表示为：\[1,0,0\]    
     
   取值B:用k维向量表示为：\[0,1,0\]  
     
   取值C:用k维向量表示为：\[0,0,1\]   
     
   各个取值可能之间的距离相同都是$\sqrt{2}$ (如AB,AC,BC)

  互信息：  
   
  互信息(Mutual Information)是信息论里一种有用的信息度量，它可以看成是一个随机变量中包含的关于另一个随机变量的信息量，或者说是一个随机变量由于已知另一个随机变量而减少的不肯定性   
    
    
  $$\sum_{x\in X}\sum_{y \in Y}p(x,y) log \; \frac{p(x,y)}{p(x) p(y)}  $$

#### 集成聚类  
  
  集成聚类，也称为共识聚类或集群聚合，旨在将多个现有的基本分区融合成一个集成的分区 。现有的工作可以大致归纳为两类根据不同的相似性水平。第一类使用效用函数来度量一致性聚类和多个基本分区在分区级别上的相似性 。第二类将基本分区的信息汇总到一个联合关联矩阵中，该矩阵计算在同一个集群中出现两个实例的次数。因此，协相关矩阵可以看作是基于实例层次的相似矩阵，在此基础上可以进行任何图划分算法进行最终聚类
    
<br>  

  
#### 思想

  相似矩阵： $S(x,y)=\frac{1}{r} \sum_{1}^r \delta (H^i(x),H^i(y))$  
    
  $\qquad\qquad\quad$ （共生矩阵）记录 样本对 共同出现的 基本分区的个数   
    
  定理：当基本分区的个数r,个数无穷大时，共生矩阵趋于稳定；此时可以用图聚类算法对共生矩阵划分。   
    
 <br>  
   
 <br>  
   
   无穷的基本分区：在原始数据集上用不同的聚类策略生成不同的分区：同一算法不同的参数，不同特征的同一算法；或者不同算法，用来生成无穷的剧本分区。
   
 本文作者针对的问题：  
   
   1、用上诉方法生成基本分区时间耗费高  
     
   2、有些任务中，没有原始数据，只有一些基本分区  
     
<br>  
  
<br>  
  
问题解决：  
  
  使用mDAE,对基本分区腐蚀，达到无穷分区的目的。  
    
   其中mDAE的输入要求是独立同分布的。但是这里的共生矩阵s不符合要求。  
     
  作者设计了一个二元矩阵b:  
    
   对r个基本分区进行1-of-k 编码  
     
   有$b(x)=<b(x)_1,\cdots,b(x)_r> \quad b(x)_i=<b(x)_{i1},\cdots,b(x)_{ik}>$   
     
   b(x)是d列，$d=\underset{r}{\sum} k_r \qquad k_r$是聚类的1-of-k 编码    
     
   $$ b(x)_{ij}= \begin{cases} 1, & H^i(x)=j \\ 0, & ohterwise \end{cases} $$    
     
   所以B是$n\times d$的矩阵；由于是 1-of-k 编码 所以独立同分布   
     
       
  <br>    
  
  $BB^T=S$  
    
      
<br>  
  
<br>  
  
无穷分区：  
  
   对B进行mDAE;得到一个W;BW^T得到一个新的B,递归以上过程直到满足递归次数L
           
<br>  

边缘去噪自编码器处理过程：  
  
  
  $S=BB^T \qquad\qquad q=[1-p,1-p,\cdots,1-p.1]$    
    
重构参数$W=E[P]E[Q]^{-1}$      

 $$ E[Q]_{\alpha,\beta}= \begin{cases}S_{\alpha,\beta}q_{\alpha}q_{\beta}, & \alpha \neq \beta \\S_{\alpha,\beta}q_{\alpha}, & \alpha = \beta\end{cases} $$  
 
 <br>
 
 $$ E[p]_{\alpha,\beta}=S_{\alpha,\beta}q_{\alpha} $$  
   
<br> 
    
求出重构参数W后，新的无穷分区的表示为：$BW^T$     
  
<br>  
  
在循环L次，之后在新的无穷分区的表示为：$BW^T$上运行K-means    
    
      
<br>  
  
<br>
