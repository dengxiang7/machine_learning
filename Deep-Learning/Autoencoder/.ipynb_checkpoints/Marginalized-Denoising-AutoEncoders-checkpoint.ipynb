### 边缘降噪自编码器    
  
  在降噪自编码器的基础上：对噪声干扰进行边缘化处理。
  
#### 单层边缘降噪自编码器（mDAE）    
  
  单层降噪自编码器（不是加噪，编码器，解码器 ；只有重构层）：原数据$x_i\;$，腐蚀数据$\hat{x_i \;}$，重构参数$W\;$,直接重构$W\hat{x_i}$，不编码  
    
  $$L=\frac{1}{n}\sum_{i=1}^n ||x_i-W\hat{x_i}||_2^2$$  
    
<br>  
  
边缘自编码器：对单层降噪自编码器重复m次(让模型更有普遍性)  
  
  
  $$L=\frac{1}{mn}\sum_{j=1}^m \sum_{i=1}^n ||x_i-W\hat{x_{ij}}||_2^2$$   
    
<br>  
  
  $$L=\frac{1}{mn}tr[(\overset{-}{X} - W\hat{X})^T(\overset{-}{X} - W\hat{X})]$$    
    
<br>  
  
  其中$X=[x_1,x_2,\cdots,x_n]$是一个版本原数据，$\overset{-}{X}=[X,X,\cdots,X]$原数据重复m次，代表m次腐蚀的原数据；$\hat{X}$代表原数据m个腐蚀版本  
    
<br>  
  
对m次重构总损失求最小化；利用最小二乘法可得解为：    
  
  $$W=PQ^{-1}$$  
    
 <br>  
   
 $$P=\overset{-}{X}\hat{X}^T \qquad Q=\hat{X}\hat{X}^T$$  
  
<br>  
  
其中求P和Q的求法：不能直接求，因为计算量很大 $\hat{X}_{d\times n*m}$    
  
<br>  

根据弱大数定律，当$m \rightarrow \infty \quad P,Q$收敛到其期望  故：$W=E[P]E[Q]^{-1}$
  
求其期望：  $S=XX^T   \qquad$Q是两个腐蚀相乘，P是一个腐蚀相乘；$S_{\alpha,\beta} \quad \alpha,\beta$是矩阵横纵下标，也就数据维度（特征） 
  
   
 $$ E[Q]_{\alpha,\beta}= \begin{cases}S_{\alpha,\beta}q_{\alpha}q_{\beta}, & \alpha \neq \beta \\S_{\alpha,\beta}q_{\alpha}, & \alpha = \beta\end{cases} $$  
 
 <br>
 
 $$ E[p]_{\alpha,\beta}=S_{\alpha,\beta}q_{\alpha} $$  
   
<br>   
  
$q=[1-p,1-p,\cdots,1-p,1]$ 是各特征不被腐蚀的概率向量，$q_{\alpha}$是特征$\alpha$不被腐蚀的概率  
  
    
<br>  
  
求解重构参数W时，只需要遍历一次数据集就行了，计算速度上提升了两个数量级。    
    
<br>  

<br>  

* 为什么叫边缘去噪自编码器：  
  
  因为在计算出W的过程中，我们并没有显式的去破坏数据，破坏数据的这一过程并不重要，所以噪声干扰这个过程是被边缘化的。
    
<br>    
    
**以上是单层的边缘去噪自编码器（mDAE）**   

<br>
     
### 用于领域适应性的边缘堆栈式去噪自编码器（mSDA）  
  
  非线性特征生成和叠加：   
    
    在学到重构W之后，进行保持W解不变情况下的特征学习，和堆叠降噪自编码器一样
    
    k层堆叠边缘去噪自编码器（h是学到的特征）：  
     
 $\qquad\qquad\qquad 1、 h^0=X  \quad$ 利用mDA学的$W^0$，输出为：$h^1=tanh(W^0X)$  
   
 $\qquad\qquad\qquad 2、 h^1=h^1 (上一层的输出)  \quad$ 利用mDA学的$W^1$，输出为：$h^2=tanh(W^2h^1)$    
   
 $\qquad\qquad\qquad\qquad\qquad\qquad \cdots \cdots \cdots\cdots \cdots \cdots\cdots \cdots \cdots$     
 
$\qquad\qquad\qquad 2、 h^i=h^i (上一层的输出)  \quad$ 利用mDA学的$W^i$，输出为：$h^{i+1}=tanh(W^ih^i)$   
  
$\qquad\qquad\qquad\qquad\qquad\qquad \cdots \cdots \cdots\cdots \cdots \cdots\cdots \cdots \cdots$    
   
$\qquad\qquad\qquad 2、 h^{k}=h^{k1} (上一层的输出)  \quad$ 利用mDA学的$W^k$，输出为：$h^{k+1}=tanh(W^kh^k)$   

  
在经过k层堆叠边缘堆栈式去噪自编码器之后学的特征为$h^k$，学得得重构参数为W  
       
<br>    
  
优缺点：  
  
  1、较强的特征学习能力。（$h^k$）    
    
 2、训练速度快（无优化算法）  
   
 3、更少中间参数 （中有中间一层前馈网络的参数）  
   
 4、缺点只能用线性表示
