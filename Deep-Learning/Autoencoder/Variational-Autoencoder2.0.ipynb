由上面可以知道：  
  
* 理论上的解码器：  
     $Z\text{~}N(0,1)$  
     $P_{\theta}(X|Z)$  
     $P_{\theta}(X|Z)\quad \Rightarrow\quad P_{\theta}(Z|X)$  
     即可以由理论的解码器$P_{\theta}(X|Z)$，推出理论上的后验概率$P_{\theta}(Z|X)$。尽管它是不可计算的。  

<br>         
         
* 事实上的编码器：  
     $P_{\phi}(Z|X) \Rightarrow\quad N(x;\mu_{\phi},\sigma^2_{\phi})$  
     由给定的样本计算出事实上的后验概率。  
       
<br>  
  
**$\color{red}{\text{我们的目标是，让事实上的后验概率}  P_{\phi}(Z|X)} \color{red}{\text{靠近理论上的后验概率}P_{\theta}(Z|X)。}\color{blue}{\text{刚好解决这个问题的方法是随机梯度变分推导}}$**  
  
<br>  
  
$ ln P(X) =\underbrace{\int P_{\phi}(Z|X)ln \frac{P_{\theta}(X,Z)}{P_{\phi}(Z|X)}dZ}_{\text{公式一:ELBO}}+\underbrace{KL(P_{\phi}(Z|X)||P_{\theta}(Z|X))}_{公式二}$
   
所以我们的目标是最大化下界公式ELBO公式即可（公式一:ELBO，是ln P(X)的下界）：$\underset {(\phi,\theta)}{arg \; max}ELBO$  
  
$\underset {(\phi,\theta)}{arg \; max}\;\; \int P_{\phi}(Z|X)ln \frac{P_{\theta}(X,Z)}{P_{\phi}(Z|X)}dZ$  
  
令：$L_{(\phi,\theta)} \;=\; \int P_{\phi}(Z|X)ln \frac{P_{\theta}(X,Z)}{P_{\phi}(Z|X)}dZ$  
  
即：$L_{(\phi,\theta)} \;=\; E_{P_{\phi}(Z|X)}[ln P_{\theta}(X,Z)-ln P_{\phi}(Z|X)]$  
  
$\qquad\quad\quad =\; E_{P_{\phi}(Z|X)}[ln \left(P_{\theta}(X|Z)P_{\theta}(Z)\right)-ln P_{\phi}(Z|X)]$  
  
$\qquad\quad\quad =\; E_{P_{\phi}(Z|X)}[ln P_{\theta}(X|Z) + ln P_{\theta}(Z)-ln P_{\phi}(Z|X)]$  
  
$\qquad\quad\quad =\; E_{P_{\phi}(Z|X)}[ln P_{\theta}(X|Z) -ln \frac{P_{\phi}(Z|X)}{P_{\theta}(Z)}]$
  
$\qquad\quad\quad =\; E_{P_{\phi}(Z|X)}[ln P_{\theta}(X|Z)] - \int P_{\phi}(Z|X) ln \frac{P_{\phi}(Z|X)}{ln P_{\theta}(Z)}dZ$  
  
$\qquad\quad\quad =\; E_{P_{\phi}(Z|X)}[ln P_{\theta}(X|Z)] -D_{ KL}( P_{\phi}(Z|X) || P_{\theta}(Z))$  
  
<br>  
  
**目标函数为：$\underset {(\phi,\theta)}{arg \; max \;} L_{(\phi,\theta)} \;=\; \underbrace{E_{P_{\phi}(Z|X)}[ln P_{\theta}(X|Z)]}_{part1}\underbrace{ -D_{ KL}( P_{\phi}(Z|X) || P_{\theta}(Z))}_{part2} $**  
  
这个公式part2中：$P_{\theta}(Z)$来自理论中的分布，我们可以选择标准正态分布$N(0,1)$  
  
对于公式part1中： 计算期望，我们要对$P_{\phi}(Z|X) \;\;\Rightarrow\;\; N(x;\mu_{\phi},\sigma^2_{\phi})$ 进行采样，**为了保留采样操作中的参数** ，我们对这里的采样选择**重参数**技巧   
  
$\qquad\qquad\qquad\qquad $即我们选择：从$\epsilon \text {~}N(0,1)$采样$\epsilon $,再转化为Z:$\;\; Z=\mu+\sigma *\epsilon $  
  
$\qquad\qquad\qquad\qquad $这样我们可以保留参数$(\phi,\theta)$的进行采样，并计算出part1，带参数的期望。  
  
对$ L_{(\phi,\theta)} $公式，中使用重参数计算出part1期望，之后的公式，**我们另记为：$\hat{L}_{(\phi,\theta)}$**  
   
<br>  
 
**之后要最大化$\underset {(\phi,\theta)}{arg \; max \;} \hat{L}_{(\phi,\theta)}$,我们就可以使用，随机梯度了**  
  
$$(\phi,\theta)^{t+1}=(\phi,\theta)^{t}+ \eta \nabla \hat{L}_{(\phi,\theta)}$$  
  
    
**求出最大化的结果，即$(\phi,\theta)$值,我们也就求出了两个概率分布，即可以得到理论上的生成模型的参数**

**$\color{red}{\text{只是在不同的迭代周期中。}\theta  和P_{\theta}(X) \text{是不同的。}} $**
